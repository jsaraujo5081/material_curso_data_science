## Tarea 9

Finalmente, se te comparte una versión simplificada de la información de 3 usuarios distintos.

```py
users = [
    [' Mike Richards', "78"],
    ['Rozalia_Luxemburg', "47"],
    ['Kevin Jones ', "41"]
]
```

Se te pide que valides algunos de los procesos de calidad que ejecutaste. En concreto:

- Elimina todos los espacios en blanco al inicio y al final de los nombres, así como cualquier guion bajo.
- Separa todos los nombres y apellidos en una lista.    
- Convierte todas las edades a números enteros.  
- Guardar los resultados en una nueva lista llamada `users_clean` que tenga la misma estructura que `users` y luego muestrala en pantalla.

````python
users = [
  [' Mike Richards', "78"],
  ['Rozalia_Luxemburg', "47"],
  ['Kevin Jones ', "41"]
]

# Definir la nueva lista
users_clean = []

# Procesar el primer usuario
user_name_1 = users[0][0].strip()
user_name_1 = user_name_1.replace('_', ' ')
user_name_1 = user_name_1.split()
user_age_1 = int(users[0][1])
  
users_clean.append([user_name_1, user_age_1])

# Procesar el segundo usuario
user_name_2 = users[1][0].strip()
user_name_2 = user_name_2.replace('_', ' ')
user_name_2 = user_name_2.split()
user_age_2 = int(users[1][1])
  
users_clean.append([user_name_2, user_age_2])

# Procesar el tercer usuario
user_name_3 = users[2][0].strip()
user_name_3 = user_name_3.replace('_', ' ')
user_name_3 = user_name_3.split()
user_age_3 = int(users[2][1])
  
users_clean.append([user_name_3, user_age_3])

# Imprimir resultado en la pantalla
print(users_clean)
````
