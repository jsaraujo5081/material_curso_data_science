# Sprint 8 - Manejo de Bases de Datos con SQL (Ejercicios)

En el presente caso de estudio profundizaremos en el uso del lenguaje **SQL** mediante la resolución de un problema de negocio ampliamente utilizado en áreas como marketing, ventas, comercio electrónico, productos digitales y gestión de clientes: el análisis de embudos. Este tipo de análisis constituye una de las aplicaciones más relevantes de la análitica de datos dentro de las organizaciones modernas, ya que por un lado permite comprender cómo evolucionan las diferentes entidades de un negocio a través de un proceso definido y cuáles son los factores que influyen en su avance o abandono; y por otro caracterizan de mejor forma a dichas entidades en términos de valor. 

Desde una perspectiva general, toda organización busca que determinadas entidades completen un conjunto de etapas que generan valor para el negocio. Dependiendo del contexto, estas entidades pueden corresponder a clientes potenciales, usuarios de una aplicación, suscriptores de un servicio, productos dentro de una cadena logística, solicitudes de crédito, estudiantes de un programa académico o cualquier otro elemento que transite por un flujo de actividades previamente establecido. El análisis de embudos tiene precisamente como objetivo estudiar ese recorrido y cuantificar el desempeño de cada una de las etapas que lo componen.

La denominación de "embudo" proviene de la representación visual que suele adoptar este tipo de análisis. A medida que las entidades avanzan por las diferentes etapas del proceso, una parte de ellas abandona o no logra completar los requisitos necesarios para continuar. Como consecuencia, el número de entidades presentes en cada fase disminuye progresivamente, generando una forma similar a la de un embudo: una base amplia en las etapas iniciales y una sección más estrecha en las etapas finales. Por ejemplo, dentro de un proceso comercial es habitual observar que un gran número de personas visualiza una campaña publicitaria, una proporción menor manifiesta interés en el producto, una cantidad aún más reducida solicita información adicional y solamente una fracción termina realizando una compra.

Este estudio permite identificar puntos críticos dentro del proceso de negocio. Una caída significativa en una etapa específica puede indicar problemas operativos, barreras para los usuarios, deficiencias en la propuesta de valor o oportunidades de mejora que afectan directamente los resultados de la organización. Por esta razón, el análisis de embudos se ha convertido en una herramienta indispensable para la optimización de procesos y la toma de decisiones basada en datos.

De manera general, el análisis de embudos suele dividirse en dos grandes componentes complementarios:

* Análisis de adquisición de entidades: Se enfoca en comprender cómo las entidades avanzan a través de las distintas etapas del embudo y cuáles son las probabilidades de conversión entre cada una de ellas. Su objetivo principal es identificar dónde se producen las mayores pérdidas y qué factores podrían estar influyendo en el comportamiento observado. En este contexto, indicadores como tasas de conversión y abandono, y volúmenes de entidades resultan fundamentales para evaluar la eficiencia del proceso. A través de estos análisis es posible detectar cuellos de botella, oportunidades de optimización y segmentos con comportamientos diferenciados, proporcionando información valiosa para mejorar el desempeño general del negocio.
* Análisis de retención de entidades: Mientras que la adquisición se centra en el avance dentro del embudo, el análisis de retención se enfoca en el comportamiento de las entidades que alcanzan las etapas finales del proceso. El interés principal radica en comprender si estas entidades continúan generando valor para la organización a lo largo del tiempo y cuál es su contribución efectiva al negocio. En el caso de clientes o usuarios, por ejemplo, no basta con analizar cuántos completan una compra o una suscripción. También resulta relevante estudiar aspectos como frecuencia de uso, recurrencia de compra, permanencia en el servicio, nivel de interacción y valor económico generado. Este enfoque permite evaluar la calidad de las conversiones obtenidas y no únicamente su cantidad neta.

En este sentido, la combinación de adquisición y retención proporciona una visión integral del ciclo de vida de las entidades dentro de una organización. Mientras la primera permite comprender cómo llegan las entidades a las etapas de valor, la segunda ayuda a determinar si dichas entidades mantienen una relación sostenible y rentable con el negocio.

Una de las aplicaciones más importantes del análisis de embudos se encuentra actualmente en los procesos de marketing y ventas. En estos contextos, las organizaciones buscan comprender el denominado viaje del cliente, es decir, el conjunto de interacciones y decisiones que una persona realiza desde el primer contacto con una marca, producto o servicio, hasta la conversión final y su posible fidelización. Analizar este recorrido permite identificar oportunidades para mejorar campañas publicitarias, optimizar canales de adquisición, incrementar conversiones y fortalecer la experiencia de los clientes.

En el presente caso de estudio utilizaremos consultas avanzadas en **SQL* para construir y analizar embudos de negocio, calculando indicadores de conversión, identificando patrones de comportamiento y generando información útil para la toma de decisiones. De esta manera, se combinarán competencias técnicas de manipulación y consulta de datos con habilidades analíticas orientadas a resolver problemas reales presentes en organizaciones modernas.

## Preparación de datos

Ejecuta el plan de acción especificado en la sección anterior. Inicia por tanto creando la columna gender_full mediante los comandos `ALTER TABLE`, `ADD COLUMN`. Una vez hecho esto, rellénala con información más semántica y comprensible mediante los comandos `UPDATE`, `SET` y `CASE...END`.

```sql
--Crear columna nueva
ALTER TABLE fac_adquisicion
ADD COLUMN gender_full VARCHAR(100);

--Inputar valores a columna nueva
UPDATE fac_adquisicion
SET gender_full =
  CASE
      WHEN gender = 'w' 
      THEN 'female'
      ELSE 'male'
  END;
```

Cambia el tipo de la columna months_retained a entero con los comandos `ALTER TABLE` y `ALTER COLUMN` y `TYPE`.

```sql
ALTER TABLE fac_adquisicion
ALTER COLUMN months_retained 
TYPE INTEGER;
```

Crea una nueva columna llamada months_retained_clean en donde se limpien los valores ausentes de months_retained.

```sql
--Crear columna nueva
ALTER TABLE fac_adquisicion
ADD COLUMN months_retained_clean INTEGER;

--Inputar valores a columna nueva
UPDATE fac_adquisicion
SET months_retained_clean =
  CASE
      WHEN months_retained IS NULL 
      THEN 0
      ELSE months_retained
  END;
```

Repite este procedimiento para el caso de package_price.

```sql
--Crear columna nueva
ALTER TABLE fac_adquisicion
ADD COLUMN package_price_clean INTEGER;

--Inputar valores a columna nueva
UPDATE fac_adquisicion
SET package_price_clean =
  CASE
      WHEN package_price IS NULL 
      THEN 0
      ELSE package_price
  END;
```

### ¿Cuál es la tasa de conversión de prospectos a suscriptores?

Para responder esta pregunta en primera instancia construye el embudo de este caso que muestre la cantidad de prospectos y además la cantidad de aquellos que se suscribieron. Utiliza el comando `CASE...END`para realizar un conteo condicionado de aquellos registros que se convirtieron en suscriptores.

```sql
SELECT
count(prospect_id) AS prospectos,
count(
    CASE 
        WHEN converted 
        THEN prospect_id 
    END
) AS suscriptores
FROM vw_adquisicion;
```

A continuación, modifica esta consulta para que se incluya además el porcentaje de suscriptores respecto al total de prospectos. Este valor se conoce como *tasa de conversión* y es el principal indicador asociado a los embudos, puesto que mide la calidad de los potenciales clientes que llegan gracias a las acciones de marketing.

```sql
SELECT
count(prospect_id) AS prospectos,
count(
    CASE 
        WHEN converted 
        THEN prospect_id 
    END
) AS suscriptores,
100.0 * count(
    CASE 
        WHEN converted 
        THEN prospect_id 
    END
) / count(*) AS tasa_conversion
FROM vw_adquisicion;
```

### ¿Que otras variables inciden en la tasa de conversión?

Cambiemos ahora de grupos relevantes para nuestro análisis y consideremos la edad de los prospectos. Crea entonces una consulta que permita evidenciar la tasa de conversión por edad.

```sql
SELECT
age,
count(prospect_id) AS prospectos,
count(
    CASE 
        WHEN converted 
        THEN prospect_id 
    END
) AS suscriptores,
100.0 * count(
    CASE 
        WHEN converted 
        THEN prospect_id 
    END
) / count(*) AS tasa_conversion
FROM vw_adquisicion
GROUP BY 
age
ORDER BY
age;
```

Es evidente que existen muchas cohortes, y extraer conclusiones es complejo por la dificultad asociada a visualizar los resultados. Por tanto genera una consulta que muestre la tasa de conversión para todos los prospectos menores a 30 años. En esta consulta incorpora una columna grupo_edad con el valor "a.Menores a 30".

```sql
SELECT
'a.Menores a 30' AS grupo_edad,
count(prospect_id) AS prospectos,
count(
    CASE 
        WHEN converted 
        THEN prospect_id 
    END
) AS suscriptores,
100.0 * count(
    CASE 
        WHEN converted 
        THEN prospect_id 
    END
) / count(prospect_id) AS tasa_conversion
FROM vw_adquisicion
WHERE age < 30;
```

Repite este procedimiento pero ahora para los siguientes grupos:

* Entre 30 y 40 años.
* Mayores o iguales a 40 años.

```sql
--Tasa de conversión entre 30 y 40 años
SELECT
'b.Entre 30 y 40' AS grupo_edad,
count(prospect_id) AS prospectos,
count(
    CASE 
        WHEN converted 
        THEN prospect_id 
    END
) AS suscriptores,
100.0 * count(
    CASE 
        WHEN converted 
        THEN prospect_id 
    END
) / count(prospect_id) AS tasa_conversion
FROM vw_adquisicion
WHERE age >= 30 AND age < 40;

--Tasa de conversión mayores o iguales a 40
SELECT
'c.Mayores o iguales a 40' AS grupo_edad,
count(prospect_id) AS prospectos,
count(
    CASE 
        WHEN converted 
        THEN prospect_id 
    END
) AS suscriptores,
100.0 * count(
    CASE 
        WHEN converted 
        THEN prospect_id 
    END
) / count(prospect_id) AS tasa_conversion
FROM vw_adquisicion
WHERE age >= 40;
```

A fin de consolidar todos estos resultados en una sola tabla, reutiliza las consultas anteriores dentro de tablas temporales (CTEs) mediante el comando `WITH`, y aplica el comando `UNION ALL` para unir las salidas de cada una.

```sql
WITH 
cte_menores_30 AS (
  SELECT
  'a.Menores a 30' AS grupo_edad,
  count(prospect_id) AS prospectos,
  count(
      CASE 
          WHEN converted 
          THEN prospect_id 
      END
  ) AS suscriptores,
  100.0 * count(
      CASE 
          WHEN converted 
          THEN prospect_id 
      END
  ) / count(prospect_id) AS tasa_conversion
  FROM vw_adquisicion
  WHERE age < 30
),

cte_entre_30_40 AS (
  SELECT
  'b.Entre 30 y 40' AS grupo_edad,
  count(prospect_id) AS prospectos,
  count(
      CASE 
          WHEN converted 
          THEN prospect_id 
      END
  ) AS suscriptores,
  100.0 * count(
      CASE 
          WHEN converted 
          THEN prospect_id 
      END
  ) / count(prospect_id) AS tasa_conversion
  FROM vw_adquisicion
  WHERE age >= 30 AND age < 40
),

cte_mayores_40 AS (
  SELECT
  'c.Mayores o iguales a 40' AS grupo_edad,
  count(prospect_id) AS prospectos,
  count(
      CASE 
          WHEN converted 
          THEN prospect_id 
      END
  ) AS suscriptores,
  100.0 * count(
      CASE 
          WHEN converted 
          THEN prospect_id 
      END
  ) / count(prospect_id) AS tasa_conversion
  FROM vw_adquisicion
  WHERE age >= 40
)

SELECT 
* 
FROM cte_menores_30
UNION ALL
SELECT 
* 
FROM cte_entre_30_40
UNION ALL
SELECT 
* 
FROM cte_mayores_40
ORDER BY
grupo_edad;
```

### ¿Cuál ha sido el retorno efectivo sobre la inversión de marketing?

Para contestar esta pregunta definamos otro indicador conocido como *ARPU: Average Revenue Per User* que mide el ingreso promedio efectivo recibido por un cliente. Para calcularlo sea entonces el precio medio mensual de suscripción $p_s$. Obtén este precio mediante una consulta de la vista de suscripciones (recuerda filtrar los meses de creación anteriores a Agosto). 

```sql
SELECT
avg(package_price) AS precio_medio
FROM vw_suscriptores
WHERE created_month <= 7;
```

El ARPU por tanto se define por la siguiente fórmula:

$$ ARPU = \sum_{m} (r_m\times p_s) = p_s\sum_{m} r_m $$

donde $r_m$ es la tasa de retención en cada uno de los $m$ meses. Calcula entonces este indicador mediante CTEs en base a tus consultas previas de retención (no hace falta que agrupes por mes de creación). 

```sql
WITH
cte_retencion AS (
  SELECT
  100.0 * count(subscriber_id) / count(subscriber_id)  AS retencion_1m,
  100.0 * count(
      CASE 
          WHEN months_retained >= 2 
          THEN subscriber_id 
      END
  ) / count(subscriber_id) AS retencion_2m,
  100.0 * count(
      CASE 
          WHEN months_retained >= 3 
          THEN subscriber_id 
      END
  ) / count(subscriber_id) AS retencion_3m,
  100.0 * count(
      CASE 
          WHEN months_retained >= 4 
          THEN subscriber_id 
      END
  ) / count(subscriber_id) AS retencion_4m,
  100.0 * count(
      CASE 
          WHEN months_retained >= 5 
          THEN subscriber_id 
      END
  ) / count(subscriber_id) AS retencion_5m,
  100.0 * count(
      CASE 
          WHEN months_retained >= 6 
          THEN subscriber_id 
      END
  ) / count(subscriber_id) AS retencion_6m
  FROM vw_suscriptores
  WHERE created_month <= 7
),

cte_precio AS (
  SELECT
  avg(package_price) AS precio_medio
  FROM vw_suscriptores
  WHERE created_month <= 7
)

SELECT
(
  SELECT 
  * 
  FROM cte_precio
) * (retencion_1m + retencion_2m + retencion_3m + retencion_4m + retencion_5m + retencion_6m) / 100.0 AS arpu
FROM cte_retencion;
```

Mediante una consulta a la vista de inversiones, calcula ahora el costo que han tenido las campañas de marketing en los meses anteriores a Agosto.

```sql
SELECT
sum(cost) AS cost
FROM vw_inversion
WHERE month <= 7;
```

Incorpora en la consulta del ARPU el costo que se ha tenido por cada uno de los suscriptores. Este indicador se conoce como *CAC: Client Acquisition Cost* y lo puedes obtener dividiendo el costo calculado antes para el total de suscriptores existentes.

```sql
WITH
cte_retencion AS (
  SELECT
  count(subscriber_id) AS suscriptores_iniciales,
  100.0 * count(subscriber_id) / count(subscriber_id)  AS retencion_1m,
  100.0 * count(
      CASE 
          WHEN months_retained >= 2 
          THEN subscriber_id 
      END
  ) / count(subscriber_id) AS retencion_2m,
  100.0 * count(
      CASE 
          WHEN months_retained >= 3 
          THEN subscriber_id 
      END
  ) / count(subscriber_id) AS retencion_3m,
  100.0 * count(
      CASE 
          WHEN months_retained >= 4 
          THEN subscriber_id 
      END
  ) / count(subscriber_id) AS retencion_4m,
  100.0 * count(
      CASE 
          WHEN months_retained >= 5 
          THEN subscriber_id 
      END
  ) / count(subscriber_id) AS retencion_5m,
  100.0 * count(
      CASE 
          WHEN months_retained >= 6 
          THEN subscriber_id 
      END
  ) / count(subscriber_id) AS retencion_6m
  FROM vw_suscriptores
  WHERE created_month <= 7
),

cte_precio AS (
  SELECT
  avg(package_price) AS precio_medio
  FROM vw_suscriptores
  WHERE created_month <= 7
),

cte_costo AS (
  SELECT
  sum(cost) AS cost
  FROM vw_inversion
  WHERE month <= 7
)

SELECT
(
  SELECT 
  * 
  FROM cte_precio
) * (retencion_1m + retencion_2m + retencion_3m + retencion_4m + retencion_5m + retencion_6m) / 100.0 AS arpu,
(
  SELECT 
  * 
  FROM cte_costo
) / suscriptores_iniciales AS cac
FROM cte_retencion;
```

Ya estamos finalizando. debemos ahora restar del ARPU el CAC. Este valor se conoce como *CLV: Customer Lifetime Value* y mide la rentabilidad efectiva que tiene cada suscriptor derivado de las acciones de marketing realizadas. Incorpóralo por tanto en la consulta.

```sql
WITH
cte_retencion AS (
  SELECT
  count(subscriber_id) AS suscriptores_iniciales,
  100.0 * count(subscriber_id) / count(subscriber_id)  AS retencion_1m,
  100.0 * count(
      CASE 
          WHEN months_retained >= 2 
          THEN subscriber_id 
      END
  ) / count(subscriber_id) AS retencion_2m,
  100.0 * count(
      CASE 
          WHEN months_retained >= 3 
          THEN subscriber_id 
      END
  ) / count(subscriber_id) AS retencion_3m,
  100.0 * count(
      CASE 
          WHEN months_retained >= 4 
          THEN subscriber_id 
      END
  ) / count(subscriber_id) AS retencion_4m,
  100.0 * count(
      CASE 
          WHEN months_retained >= 5 
          THEN subscriber_id 
      END
  ) / count(subscriber_id) AS retencion_5m,
  100.0 * count(
      CASE 
          WHEN months_retained >= 6 
          THEN subscriber_id 
      END
  ) / count(subscriber_id) AS retencion_6m
  FROM vw_suscriptores
  WHERE created_month <= 7
),

cte_precio AS (
  SELECT
  avg(package_price) AS precio_medio
  FROM vw_suscriptores
  WHERE created_month <= 7
),

cte_costo AS (
  SELECT
  sum(cost) AS cost
  FROM vw_inversion
  WHERE month <= 7
)

SELECT
(
  SELECT 
  * 
  FROM cte_precio
) * (retencion_1m + retencion_2m + retencion_3m + retencion_4m + retencion_5m + retencion_6m) / 100.0 AS arpu,
(
  SELECT 
  * 
  FROM cte_costo
) / suscriptores_iniciales AS cac,
(
  SELECT 
  * 
  FROM cte_precio
) * (retencion_1m + retencion_2m + retencion_3m + retencion_4m + retencion_5m + retencion_6m) / 100.0 - (
  SELECT 
  * 
  FROM cte_costo
) / suscriptores_iniciales AS clv
FROM cte_retencion;
```
