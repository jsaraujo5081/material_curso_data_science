## Preparación de datos

Implementa el plan de acción que definiste de forma organizada.

```python
# Cambiar nombres de columnas a formato snake_case
df_games.columns = [nombre.lower().replace(" ","_") for nombre in df_games.columns]
df_games.info()
```

```python
# Eliminar valores perdidos en columna name, genre y year_of_release
df_games = df_games.dropna(subset = ["name", "genre", "year_of_release"])
df_games.info()
```

```python
# Cambiar el tipo de la variable year_of_release
df_games["year_of_release"] = df_games["year_of_release"].astype(int)
df_games.info()
```

```python
# Crear columna con grupos de plataformas
def fun_plataforma (x):
    if x in ["3DS","DS","GB","GBA","GC","N64","NES","SNES","Wii","WiiU"]:
        return "Nintendo"
    elif x in ["PC","PCFX"]:
        return "PC"
    elif x in ["PS","PS2","PS3","PS4","PSP","PSV"]:
        return "PlayStation"
    elif x in ["X360","XB","XOne"]:
        return "XBox"
    else:
        return "Other"

df_games["platform_group"] = df_games["platform"].apply(fun_plataforma)
df_games["platform_group"].value_counts()
```

```python
# Sustituir "tbd" por valor perdido numérico en user_score
df_games["user_score"] = df_games["user_score"].replace("tbd",np.nan)
df_games["user_score"].unique()
```

```python
# Transformar user_score a flotate
df_games["user_score"] = df_games["user_score"].astype(float)
df_games.info()
```

```python
# Multiplicar por 10 valor de user_score
df_games["user_score"] = df_games["user_score"] * 10
df_games[["critic_score","user_score"]].describe().round(2)
```

```python
# Cambiar rating por valores más semánticos
def fun_rating (x):
    if x == "E" or x == "K-A":
        return "Everyone"
    elif x == "T":
        return "Teens"
    elif x == "M":
        return "Mature"
    elif x == "E10+":
        return "Everyone 10y+"
    elif x == "EC":
        return "Early Childhood"
    elif x == "RP":
        return "Pending"
    elif x == "AO":
        return "Adults Only"
    else:
        return "Pending"

df_games["rating"] = df_games["rating"].apply(fun_rating)
df_games["rating"].value_counts()
```

```python
# Generar columna total_sales
df_games["total_sales"] = df_games[["na_sales","eu_sales","jp_sales","other_sales"]].sum(axis = 1)
df_games["total_sales"].describe().round(2)
```

```python
# Eliminar duplicados detectados
df_games = df_games.drop_duplicates(subset = ["name","platform","year_of_release"])
print(df_games.duplicated(subset = ["name","platform","year_of_release"]).sum())
```

### ¿Existen diferencias significativas en el promedio de ventas anuales entre Nintendo y Playstation?

Para esta inferencia vamos a trabajar en dos etapas: inicialmente desarrollaremos una prueba de varianzas, cuyo resultado utilizaremos como insumo en la prueba de medias. Al respecto, preimero deberás usar un test de hipótesis distinto a los vistos hasta ahora, en el que se busque constrastar estadísticos de dispersión. Una vez comprobado si la variabilidad de las ventas de Nintendo es distinta a la de Playstation, incorporaremos este resultado como un argumento en la prueba de medias requerida para contestar la pregunta.

Ante esto, En **Scipy** se tienen las siguientes opciones de test disponibles para contraste de varianzas:

* Bartlett: Asume que los datos muestrales distribuyen de forma normal (función `bartlett`).
* Levene: Asume que los datos muestrales NO distribuyen de forma normal (función `levene`).

Dado que no hemos realizado ninguna comprobación asociada a la distribución de los datos, ejecuta ambas con un nivel de significancia del 10%.

**HIPÓTESIS NULA Y ALTERNATIVA**

| Hipótesis | Postulado |
| ---- | ---- |
| Nula | La varianza de las ventas de Nintendo es IGUAL a la de las ventas PlayStation. |
| Alternativa | La varianza de las ventas de Nintendo es DIFERENTE a la de las ventas PlayStation. |

```python
# Extraer muestras para el análisis
m_Nintendo = sales_per_platform["Nintendo"]
m_Playstation = sales_per_platform["PlayStation"]
pd.concat([m_Nintendo.describe(), m_Playstation.describe()], axis = 1).round(1)
```

```python
# Verificar si las varianzas son iguales o no con una prueba de Bartlett
p_bartlett = st.bartlett(m_Nintendo,m_Playstation).pvalue
print(f'Valor p: {p_bartlett:.1%}')
if p_bartlett > 0.1:
    print("No se rechaza la hipotesis nula de que las varianzas sean iguales")
else:
    print("Sí se rechaza la hipotesis nula de que las varianzas sean iguales")
```

```python
# Verificar si las varianzas son iguales o no con una prueba de Levene
p_levene = st.levene(m_Nintendo,m_Playstation).pvalue
print(f'Valor p: {p_levene:.1%}')
if p_levene > 0.1:
    print("No se rechaza la hipotesis nula de que las varianzas sean iguales")
else:
    print("Sí se rechaza la hipotesis nula de que las varianzas sean iguales")
```

Existe suficiente evidencia para afirmar que las dispersiónes de las ventas de Nintendo y Playstation son diferentes. Por tanto, realiza ahroa el test de medias incorporando el argumento `equal_var` correspondiente.

**HIPÓTESIS NULA Y ALTERNATIVA**

| Hipótesis | Postulado |
| ---- | ---- |
| Nula | El promedio de las ventas de Nintendo es IGUAL a la de las ventas PlayStation. |
| Alternativa | El promedio de las ventas de Nintendo es DIFERENTE a la de las ventas PlayStation. |

```python
p_ttest = st.ttest_ind(m_Nintendo,m_Playstation, equal_var = False).pvalue
print(f'Valor p: {p_ttest:.1%}')
if p_ttest > 0.1:
    print("No se rechaza la hipotesis nula de que las medias de venta anual son iguales")
else:
    print("Sí se rechaza la hipotesis nula de que las medias de venta anual son iguales")
```
