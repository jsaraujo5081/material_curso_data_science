# Sprint 11 - Modelos de Regresión (Sesiones)

En el presente caso de estudio nos enfocaremos en una de las áreas más importantes del aprendizaje supervisado: los modelos de **regresión**. Como se mencionó en proyectos anteriores, este tipo de modelos tiene como objetivo predecir o estimar valores numéricos a partir del conjunto de atributos. A diferencia de los problemas de clasificación, donde el resultado esperado corresponde a una categoría o clase, en los problemas de **regresión** el interés radica en estimar magnitudes de naturaleza cuantitativa.

Los modelos de **regresión** tienen una enorme relevancia en el ámbito empresarial, científico e industrial, ya que permiten anticipar comportamientos futuros y apoyar la toma de decisiones mediante pronósticos basados en información histórica. Por ejemplo, una empresa puede utilizar modelos de regresión para estimar la demanda futura de sus productos, una entidad financiera para proyectar ingresos asociados a sus clientes, o una organización logística para prever tiempos de entrega y necesidades operativas. En todos estos casos, la capacidad de realizar estimaciones confiables puede traducirse en ventajas competitivas y mejoras significativas en la eficiencia de los procesos.

Para abordar este tipo de problemas estudiaremos dos algoritmos fundamentales dentro del aprendizaje automático supervisado:

* **Regresión Lineal**: uno de los modelos más clásicos y ampliamente utilizados en estadística y ciencia de datos. Su objetivo es modelar la relación existente entre una variable objetivo y un conjunto de atributos explicativos mediante una función lineal. Además de ser un algoritmo relativamente sencillo de implementar e interpretar, constituye la base conceptual de numerosos métodos más avanzados utilizados actualmente en analítica predictiva.
* **Vecinos Más Cercanos (KNN)**: un algoritmo basado en similitud que realiza predicciones utilizando el comportamiento observado en registros históricos similares. A diferencia de la regresión lineal, no asume una forma matemática específica para la relación entre variables, lo que le permite adaptarse a patrones más complejos presentes en los datos. Su funcionamiento resulta especialmente útil para comprender conceptos relacionados con *distancias*.

Sin embargo y como ya lo hemos venido destacando, la construcción de modelos predictivos efectivos no depende únicamente de la selección del algoritmo. Una parte fundamental del proceso consiste en garantizar que los datos utilizados para el entrenamiento presenten condiciones adecuadas para que el modelo pueda aprender patrones representativos y generalizables. Por esta razón, en este caso continuaremos profundizando en diversas técnicas de ingeniería de datos:

* En primer lugar, abordaremos el tratamiento de atributos no numéricos por variabilidad, una técnica orientada a identificar variables que aportan poca o ninguna información al proceso predictivo. En muchos conjuntos de datos existen atributos cuyos valores muestran escasa variación entre observaciones, por lo que su capacidad para explicar diferencias en la variable objetivo resulta limitada. Detectar, transformar o eliminar estos casos puede simplificar los modelos y mejorar su desempeño.
* Posteriormente, estudiaremos el tratamiento de valores atípicos multidimensionales, un aspecto especialmente importante en problemas de regresión. Los valores atípicos o anomalías pueden afectar significativamente el comportamiento de ciertos algoritmos, generando estimaciones poco representativas o reduciendo la capacidad de generalización del modelo. A diferencia de los métodos univariados tradicionales, el análisis multidimensional permite identificar observaciones inusuales considerando simultáneamente múltiples variables, proporcionando una visión más completa de posibles anomalías dentro del conjunto de datos.
* También trabajaremos con técnicas de escalamiento mediante **normalización**, las cuales permiten homogeneizar las magnitudes de las variables numéricas. Dado que algunos algoritmos son sensibles a las diferencias de escala entre atributos, la normalización ayuda a evitar que determinadas variables ejerzan una influencia desproporcionada durante el entrenamiento del modelo. Este proceso resulta particularmente importante en algoritmos basados en distancias, como el KNN que utilizaremos.
* En la etapa de modelamiento, introduciremos una de las técnicas más importantes para la evaluación y optimización de modelos predictivos: la **validación cruzada**. Este método permite estimar de manera más robusta el desempeño esperado de un modelo utilizando diferentes particiones de los datos disponibles. Además de proporcionar una evaluación más confiable de la capacidad predictiva del algoritmo, la validación cruzada ayuda a detectar y controlar problemas de sobreajuste que ya habíamos mencionado.
* Finalmente, vamos a presentar métodos de optimización de modelos basados en la selección estadística de atributos. En muchas ocasiones, el uso de muchos variables predictoras puede dificultar que el algoritmo encuentre aquellos patrones relevante para generar sus pronósticos, reduciendo su efectividad e interpretabilidad. Por tanto, conviene incorporar un criterio estadístico que garantice que se están tomando en consideración únicamente a aquellos atributos suficientemente relevantes.

### Balanceo de atributos no numéricos

Recordemos nuestro simil y adicionemos un nuevo criterio de aprendizaje a los que ya conocemos:

*Para que nuestro niño aprenda mejor a sumar, debemos presentarle ejercicios variados, es decir, intentar que las sumas que resuelva contengan distintos números como sumandos y que no se repitan con tanta frecuencia respecto a otros. Si siempre resuelve operaciones que incluyan el número 5 por ejemplo, no será capaz de dar la con respuesta correcta de $7 + 8$.*    

En este sentido, notemos que para que el modelo cuente con el mencionado criterio, es importante que exista un buen balanceo de todos los atributos que deseamos incorporar como insumo del algoritmo a implementar.

Recuerda que ya estudiamos sobre cómo balancear las clases de la variable objetivo, lo cual nos permitía generar confiabilidad en la evaluación de asertividad de un modelo de clasificación. Ahora bien, extendiendo este concepto hacia los atributos, se dice que una variable numérica está balanceado si su nivel de **dispersión** es suficientemente alto; y cuando la misma es no numérica, si su nivel de **concentración** es bajo. 

Al respecto, existen muchos indicadores para estudiar la concentración de una variable no numérica, sin embargo por ahora nos enfocaremos en los dos más utilizados:

* Razón de frecuencia ($R_f$): Mide el número de veces que la frecuencia de la clase del atributo más común ($f_m$) representa frente a la frecuencia promedio de las otras clases ($f_p$). En este sentido, si la razón de frecuencia es alta se dice que existe concentración en el atributo, y por tanto desbalanceo.

$$ R_f = \frac{f_{m}}{f_{p}} $$

* Razón de unicidad ($R_u$): Mide la relación entre la cantidad de clases existentes del atributo ($q_a$) y el total de registros observados en el dataset ($n$). En este sentido, si la razón de unicidad es baja se dice que existe concentración en el atributo, y por tanto desbalanceo.

$$ R_u = \frac{q_a}{n}$$

Construye una función que estime estos indicadores para cualquier atributo no numérico que reciba como entrada.

```python
def concentracion (columna):

    # Generar tabla de frecuencia de columna
    tabla_frec = columna.value_counts().sort_values(ascending = False)

    # Definir los ratios relevantes
    r_f = tabla_frec.iloc[0]/tabla_frec.iloc[1:].median()
    r_u = len(tabla_frec)/tabla_frec.sum()

    return r_f, r_u
```

Aplica la función creada a la columna tipo del dataset.

```python
rf_tipo, ru_tipo = concentracion(df_posts["tipo"])
print(f"Razon de frecuencia: {rf_tipo:.1f}")
print(f"Razon de unicidad: {ru_tipo:.1%}")
```

Los resultados alcanzados dan cuenta que este atributo presenta desbalanceo, lo cual es coherente porque gran parte de las publicaciones en el dataset son "Fotos" (426 de 500). Ante esto, existen 3 posibles rutas de acción en cuanto al proceso de ingeniería:

* Si la razón de frecuencia es mayor a 5 y la razón de unicidad es menor a 3%, entonces el atributo debería eliminarse dado que su desbalanceo es significativamente alto y por tanto no aportará nada en la construcción de un modelo predictivo. 
* Si el ratio de frecuencia es mayor a 5 pero la razón de unicidad es mayor o igual a 3%, entonces el atributo debe transformarse mediante un proceso de agrupamiento de sus valores que generen un mayor balanceo. 
* En cualquier otro escenario, el atributo se puede considerar suficientemente balanceado.

Modifica la función creada para que despliegue un mensaje pertinente en base a estas condiciones, y comprueba su funcionamiento con la variable tipo.

```python
# Modificar funcion para estimar concentración
def concentracion (columna):

    # Generar tabla de frecuencia de columna
    tabla_frec = columna.value_counts().sort_values(ascending = False)

    #Definir los ratios relevantes
    r_f = tabla_frec.iloc[0]/tabla_frec.iloc[1:].median()
    r_u = len(tabla_frec)/tabla_frec.sum()

    # Mostrar decision de ingeniería
    if r_f > 5 and r_u < 0.03:
        print("ELIMINAR ATRIBUTO")
    elif r_f > 5:
        print("AGRUPAR ATRIBUTO")
    else:
        print("MANTENER ATRIBUTO")

    return r_f, r_u
```

### Eliminación de atípicos multidimensionales

Considera ahora este nuevo criterio de aprendizaje para nuestro niño que aprende a sumar:

*Para que el niño sume mejor, los ejercicios que le entreguemos a resolver deben ser desafiantes pero evitando siempre presentarle casos que le sean imposibles de resolver por contener aspectos ajenos a sus capacidades. Por ejemplo, vale entregarle sumas como $65 + 87$ que contienen números naturales, pero no aportaría en su aprendizaje enviarle a resolver $(4 + 5i) + e^{i}$, que contiene expresiones complejas.*

Pues bien, lo anterior encaja perfectamente con el concepto de datos atípicos. Hasta ahora hemos estudiado estos valores en columnas dentro de un dataset de forma individualizada, es decir, identificándolos y de ser el caso ajustándolos una columna a la vez. Sin embargo, cuando se trabaja con tablas multidimensionales este proceso puede ser extremadamente tedioso, por lo que conviene contar con un mecanismo eficiente para identificar *outliers* aprovechando la potencial interrelación de variables.

Con este propósito, vamos a introducir un método matemático basado en el álgebra lineal y llamado **reducción dimensional**. Dicho método busca transformar un dataset con $D$ columnas numéricas en otro de tamaño $d$ tal que $d<D$, y sin que esto implique pérdida de información relevante.

Para un mejor entendimiento de la utilidad asociada a la reducción dimensional, notemos que en nuestra tabla existen 4 atributos numéricos cuya visualización conjunta es prácticamente imposible considerando nuestra “realidad” de 3 dimensiones espaciales (en efecto, no es factible generar ningún gráfico relativamente sencillo con 5 variables). Por tanto, utilicemos este método para primero reducir estas 4 dimensiones a solamente 2, y a partir de aquí poder identificar cuáles son potencialmente datos atípicos.

Extrae en un nuevo dataset las 4 variable numéricas referidas.

```python
variables = ["alcance","impresiones","usuarios","clicks"]
df_num = df_posts[variables]
df_num.info()
```

Una forma de realizar una reducción dimensional es mediante el algoritmo de **Análisis de Componentes Principales** (PCA por sus siglas en inglés). Si bien aún no vamos a entrar en mayor detalle técnico del algoritmo mencionado, en este [video](https://www.youtube.com/watch?v=HMOI_lkzW08) se da una explicación resumida de cómo funciona. En siguientes casos de estudio vamos a profundizar acerca de él y cómo generarlo.

Por ahora, transforma el dataset de variables numéricas con la función `PCA` y su método `fit`, y define el argumento `n_componentes = 2` ya que queremos solamente 2 dimensiones.

```python
pca = PCA(n_components = 2)
pca.fit(df_num)
df_pca = pd.DataFrame(
    pca.transform(df_num),
    columns = ["PC1","PC2"]
)
df_pca.info()
```

## Creación de modelo base

En esta oportunidad vamos a presentar dos diferentes algoritmos para aprendizaje de regresión. El primero y más común es el de **Regresión Lineal**, y el segundo es el de **Vecinos más Cercanos (KNN)**.

### Modelo base con regresión lineal

El algoritmo de regresión lineal busca pronosticar la variable objetivo $y$ mediante un proceso de linearización de los atributos tal que

$$ y = \sum_{x\in X} w_{x}x = w_1 x_1 + w_2 x_2 + ... $$

donde $w_x$ son pesos o coeficientes de los distintos atributos $x$, y corresponden a parámetros que se estiman en el proceso de aprendizaje por lo que no podemos manipularlos. En todo caso, este algoritmo cuenta con un hiperparámetro importante, y es la existencia de un intercepto ($w_0$) en la ecuación lineal antes descrita, por lo que en caso de definir su existencia la misma queda expresada por:

$$ y = w_0 + \sum_{x\in X} w_{x}x = w_0 + w_1 x_1 + w_2 x_2 + ... $$

Este algoritmo presenta como principal ventaja que es extremadamente sencillo de generar e interpretar. Sin embargo, esa simpleza también conlleva que no pueda ser optimizado más allá de la ya mencionada definición de existencia del intercepto. Además vale señalar que para un correcto funcionamiento del algoritmo es importante que se cumplan algunas condiciones a nivel de los atributos y la variable objetivo:

* **Linealidad**: Los atributos y la variable objetivo deben poder relacionarse de manera lineal.
* **No multicolinealidad**: Los atributos no deberían estar relacionados estadísticamente entre sí.
* **Normalidad**: La variable objetivo debería seguir una distribuir aproximadamente Normal.

Trataremos estos aspectos en próximos proyectos, pero por ahora supondremos que se cumplen satisfactoriamente.

Crea y entrena un modelo base con la función `LinearRegression` en donde se incluya un intercepto mediante el argumento `fit_intercept = True`.

```python
# Crear modelo
mod_breg = LinearRegression(fit_intercept = True)

# Entrenar modelo
mod_breg.fit(X, y)
```

Por ahora evalúa el rendimiento de este modelo, para lo cual calcula las métricas *R2*, *MAE* y *RMSE* para este modelo. Puedes emplear las funciones `metrics.r2_score`, `metrics.mean_absolute_error` y `metrics.mean_squared_error`.

```python
# Generar predicciones
y_pred = mod_breg.predict(X)

# Calcular métricas
r2_breg = metrics.r2_score(y, y_pred)
mae_breg = metrics.mean_absolute_error(y, y_pred)
rmse_breg = np.sqrt(metrics.mean_squared_error(y, y_pred))

print(f"R2: {r2_breg:.1%}")
print(f"MAE (como % del valor medio): {mae_breg / np.mean(y):.1%}")
print(f"RMSE (como % del valor medio): {rmse_breg / np.mean(y):.1%}")
```

Entendamos un poco más que nos dicen estas medidas. Considerando que existen $n$ observaciones, el coeficiente de determinación (*R2*) del modelo se lo define por la siguiente fórmula:

$$ R^2 = 1 - \frac{\sum_n (y - y_{pred})^2}{\sum_n (y - \bar y)^2}$$

Si todos los valores predichos son igual a los valores observados entonces se cumple que $R^2 = 1$, y cualquier situación distinta hará que $R^2 < 1$. Por tanto, un modelo será mejor a medida que el coeficiente de determinacion esté más próximo a 1.

Por su parte el error absoluto medio (*MAE*) se define por la siguiente fórmula:

$$ MAE = \frac{\sum_n |y - y_{pred}|}{n} $$

Por consiguiente, si este índice es cercano a 0 el modelo será mejor.

Finalmente la raiz del error cuadrático medio (*RMSE*) se lo define por la siguiente fórmula:

$$ RMSE = \sqrt{\frac{\sum_n (y - y_{pred})^2}{n}} $$

E igual que en el indice anterior, si el *RMSE* es cercano a 0 el modelo será mejor.

Ahora bien, para evaluar nuestro modelo de forma más rigurosa vamos a utilizar una nueva técnica conocida como **validación cruzada**, la cual permite un control más robusto del sobreajuste a través de la determinación objetiva de la asertividad que potencialmente se puede alcanzar ante distintos escenarios reales.

Dividamos a nuestros datos ya no solamente en 2 subconjuntos de entrenamiento y prueba, sino que ahora particionémoslo en $c$ grupos. Una vez hecho esto, a través de iteraciones tomemos como conjunto de validación a cada una de esas $c$ partes y evaluemos el rendimiento del modelo creado. Al final de este proceso tendremos una simulación exhaustiva de su funcionamiento con $c$ distintos resultados a nivel de métricas de asertividad.

Modifica entonces el modelo base con regresión lineal de forma que aprenda con validación cruzada y ya no con `fit`. Utiliza para esto la función `cross_validate` 15 grupos y tomando como métrica referencia al coeficiente de determinación, dado que es el de uso más común en este tipo de modelos. Emplea para esto los argumentos `cv = 15` y `scoring = "r2"`.

```python
# Crear modelo base
mod_breg = LinearRegression(fit_intercept = True)

# Evaluar con validación cruzada
val_cruzada = cross_validate(
    mod_breg,
    X, y,
    cv = 15,
    scoring = "r2"
)
```

Del resultado obtenido visualiza el elemento `test_score` que corresponde a los 15 valores de R2 obtenidos.

```python
# Visualizar resultados de la validación
plt.figure(figsize = [6,4])
plt.plot(
    range(15),
    val_cruzada["test_score"],
    marker = ".",
    markersize = 15,
    linestyle = ":",
    color = "dodgerblue" 
)
plt.grid(linestyle = ":")
plt.xticks(ticks = range(15), labels = range(1,16))
plt.show()
```

Guarda el valor mediano de estos resultados para tener una referencia típica de la asertividad de este modelo.

```python
r2_breg = np.median(val_cruzada["test_score"])
print(f"R2 con validacion cruzada: {r2_breg:.1%}")
```

### Modelo base con vecinos más cercanos

El algoritmo de **Vecinos más cercanos (KNN)** es un modelo de predicción bastante sencillo en su lógica, la cual consiste en predecir un valor de la variable objetivo a partir del promedio de otras $k$ observaciones "similares". Por ejemplo, si la mayoría de los posteos sobre productos con alto alcance y usuarios, publicadas en marzo los lunes a las 7 am tiene 100 interacciones como media, es esperable que un nuevo posteo con estas mismas características tenga un número similar de ellas.  

Pues bien, para determinar esta similitud entre observaciones desde una perspectiva más técnica, se toma en consideración el concepto de distancia $d$, la cual está normalmente definida para dos observaciones $a$ y $b$ de la siguiente forma:

$$ d(a,b) = \left(\sum_{x\in X}|x_a-x_b|^p\right)^{1/p} $$

donde $p$ corresponde a un argumento que determina el tipo de distancia a utilizar. Notemos que si $p = 2$ la distancia es de tipo "euclideo":

$$ d_e(a,b) = \sqrt{\sum_{x\in X}|x_a-x_b|^2} $$

Y si $p = 1$ la distancia es de tipo "manhattan":

$$ d_m(a,b) = \sum_{x\in X}|x_a-x_b| $$

La principal ventaja de este algoritmo es que no requiere del cumplimiento de ningún supuesto asociado a los atributos y a la variable objetivo, como sí sucede con la **Regresión lineal**. Sin embargo, su adecuado funcionamiento depende exclusivamente de que exista una suficiente variabilidad en los datos como para que el modelo a implementar sea capaz de identificar patrones de similitud consistentes.

Visto esto, crea un modelo base con la función `KNeighborsRegressor` y evalúalo con la misma validación cruzada que antes. Utiliza distancia euclideana y 3 vecinos como hiperparámetros iniciales del modelo, esto es, incluye como argumentos `n_neighbors = 3` y `p = 2`.

```python
# Crear modelo base
mod_bknn = KNeighborsRegressor(
    n_neighbors = 3, 
    p = 2
)

# Evaluar con validación cruzada
val_cruzada = cross_validate(
    mod_bknn,
    X, y,
    cv = 15,
    scoring= "r2"
)

# Visualizar resultados de la validación
plt.figure(figsize = [6,4])
plt.plot(
    range(15),
    val_cruzada["test_score"],
    marker = ".",
    markersize = 15,
    linestyle = ":",
    color = "darkgreen"
)
plt.grid(linestyle = ":")
plt.xticks(ticks = range(15), labels = range(1,16))
plt.show()
```

Guarda el valor mediano de estos resultados para tener una referencia típica de la asertividad de este modelo.

```python
r2_bknn = np.median(val_cruzada["test_score"])
print(f"R2 con validacion cruzada: {r2_bknn:.1%}")
```

Al comparar los dos modelos base, parece que aquel que utiliza **Regresión lineal** tiene mejores resultados. Veamos si existen cambios luego de los procesos de optimización.

## Optimización de modelos

### Ajuste de hiperparámetros con validación cruzada

Más allá del control de sobreajuste, la **validación cruzada** igualmente puede servirnos para optimizar los hiperparámetros de los algoritmos utilizados. Define para esto una función que genere una validación cruzada con 15 grupos de prueba dado un modelo y que devuelva el *R2* mediano de las iteraciones realizadas. 

```python
def val_cruzada (modelo, X, y):

    evaluacion = cross_validate(
        modelo,
        X,
        y,
        cv = 15,
        scoring = "r2"
    )
    r2_cv = np.median(evaluacion["test_score"])

    return r2_cv 
```

Aplica esta función para determinar si el algoritmo de **regresión lineal** funciona mejor con o sin intercepto.

```python
# Definir modelos
mod_reg_cint = LinearRegression(fit_intercept = True)
mod_reg_sint = LinearRegression(fit_intercept = False)

# Evaluar con validación cruzada
r2_reg_cint = val_cruzada(mod_reg_cint, X, y)
r2_reg_sint = val_cruzada(mod_reg_sint, X, y)

print(f"R2 regresión lineal con intercepto: {r2_reg_cint:.1%}")
print(f"R2 regresión lineal sin intercepto: {r2_reg_sint:.1%}")
```

Parece que en este caso, el excluir el intercepto en la regresión lineal brinda una mayor asertividad en el modelo en un entorno "real".

Ahora aplica la función para determinar los hiperparámetros óptimos en el algoritmo **KNN**.

```python
# Definir rangos de vecinos "k" y valores "p" a evaluar
k_posibles = range(2,16)
p_posibles = range(1,11)

# Ejecutar evlauaciones con validacion cruzada 
r2_iter = []
k_iter = []
p_iter = []

for v in k_posibles:
    for p in p_posibles:

        # Guardar caso evaluado
        k_iter.append(v)
        p_iter.append(p)

        # Crear modelo KNN
        modelo = KNeighborsRegressor(
            n_neighbors = v, 
            p = p
        )

        # Evaluar
        r2 = val_cruzada(modelo, X, y)
        r2_iter.append(r2)

# Organizar resultados alcanzados en dataframe
resultados_knn = pd.DataFrame(dict(
    k = k_iter,
    p = p_iter,
    r2 = r2_iter
))
```

Muestra la combinación óptima de hiperparámetros en el modelo KNN y el *R2* alcanzado.

```python
r2_knn = resultados_knn["r2"].max()
hp_opt_knn = resultados_knn.loc[resultados_knn["r2"] == r2_knn,["k","p"]]
print("Hiperparámetros óptimos KNN:")
print(hp_opt_knn)
print()
print(f"R2 KNN óptimo: {r2_knn:.1%}")
```

Si bien hay una mejora significativa respecto al modelo base con este algoritmo de aprendizaje, aún no se alcanzaron los resultados de asertividad del modelo con **Regresión lineal**, que valga decir, tampoco son ideales.

Con 14 atributos de mayor correlación y con intercepto este modelo alcanza un R2 de 28.2%. Busca ahora el número de atributos y los hiperparámetros óptimos para el modelo con **KNN** para mejorar aún más este resultado.

```python
a_lista = []
k_lista = []
p_lista = []
r2_knn_lista = []

for i in range(1,len(indices) + 1):

    # Definir dataset de atributos
    variables = indices[:i]
    X_new = X[variables]

    # Definir rangos de vecinos "k" y valores "p" a evaluar en iteracion
    k_posibles = range(2,16)
    p_posibles = range(1,11)

    # Ejecutar evaluaciones con validacion cruzada en iteracion
    r2_iter = []
    k_iter = []
    p_iter = []
    a_iter = []

    for v in k_posibles:
        for p in p_posibles:

            # Guardar caso evaluado
            a_iter.append(i)
            k_iter.append(v)
            p_iter.append(p)

            # Crear modelo KNN
            modelo = KNeighborsRegressor(
                n_neighbors = v, 
                p = p
            )

            # Evaluar
            r2 = val_cruzada(modelo, X_new, y)
            r2_iter.append(r2)

    # Organizar resultados alcanzados en iteracion 
    a_lista.extend(a_iter)
    k_lista.extend(k_iter)
    p_lista.extend(p_iter)
    r2_knn_lista.extend(r2_iter)

# Convertir resultados en dataset
df_r2_knn = pd.DataFrame(dict(
    num_atributos = a_lista,
    k = k_lista,
    p = p_lista,
    r2 = r2_knn_lista
))
```

```python
# Determinar el número de atributos y los hiperpametros de mejor rendimiento      
r2_max_knn = df_r2_knn["r2"].max()
print(df_r2_knn.query("r2 == @r2_max_knn"))
```

Con 12 atributos, 5 vecinos y una distancia euclidea, el modelo con **KNN** alcanza un coeficiente de determinación de 38.9%, superando incluso al resultado alcanzado con **Regresión lineal**. No es excelente pero es una mejora considerable respecto a nuestro estado inicial.
