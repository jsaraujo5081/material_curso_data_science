## Agrupamiento de datos

### Ejercicio 1

Muestra el precio promedio, máximo y mínimo para cada tipo de vehículo usando `groupby`.

```python
# Resolver mediante el metodo describe
(
    df
    .groupby("VehicleType")
    ["Price"].describe()
    [["mean","max","min"]]
)
```

```python
# Resolver mediante el metodo agg
(
    df
    .groupby("VehicleType")
    .agg(
        mean = ("Price", "mean"),
        max = ("Price", "max"),
        min = ("Price", "min")
    )
)
```

### Ejercicio 2

Muestra en una sola tabla resumen el precio promedio, el kilometraje mediano y la cantidad de modelos por marca.

```python
(
    df
    .groupby("Brand")
    .agg(
        precio_prom = ("Price","mean"),
        km_med = ("Kilometer","median"),
        models_q = ("Model","nunique")
    )
)
```
