## Concatenación de dataframes

Considera los siguientes tres datasets de ejemplo:

```py
df1 = pd.DataFrame(
    {'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']},
index = [0,1,2,3])

df2 = pd.DataFrame(
    {'A':['A4','A5','A6','A7'],
    'B':['B4','B5','B6','B7'],
    'C':['C4','C5','C6','C7'],
    'D':['D4','D5','D6','D7']},
index = [0,1,2,3])

df3 = pd.DataFrame(
    {'A':['A8','A9','A10'],
    'B':['B8','B9','B10'],
    'C':['C8','C9',None],

```

````python
import pandas as pd

df1 = pd.DataFrame(
    {'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']},
index = [0,1,2,3])

df2 = pd.DataFrame(
    {'A':['A4','A5','A6','A7'],
    'B':['B4','B5','B6','B7'],
    'C':['C4','C5','C6','C7'],
    'D':['D4','D5','D6','D7']},
index = [0,1,2,3])

df3 = pd.DataFrame(
    {'A':['A8','A9','A10'],
    'B':['B8','B9','B10'],
    'C':['C8','C9',None],
    'D':['D8','D9',None]},
index = [0,1,2])
````

### Ejercicio 1

Concatena df1, df2 y df3 por filas utilizando el método `concat`.

````python
resultado = pd.concat([df1, df2, df3])
print(resultado)
````

### Ejercicio 2

Concatena df1, df2 y df3 por columnas usando el método `concat`.

````python
resultado = pd.concat([df1, df2, df3], axis = 1)
print(resultado)
````

## Unión de dataframes

### Ejercicio 1

Considera los siguientes dataframes:

```py
dataset_1 = pd.DataFrame({
    'key':['K0','K1','K2','K3'],
    'A':['A0', 'A1', 'A2', 'A3'],
    'B':['B0', 'B1', 'B2', 'B3'],
})

dataset_2 = pd.DataFrame({
    'key':['K0','K1','K2','K3'],
    'A':['A4', 'A5', 'A6', 'A7'],
    'B':['B4','B5','B6','B7'],
})
```

Usa el método `merge` para unir los datasets con un *inner join*, y usando la llave key.

````python
dataset_1 = pd.DataFrame({
    'key':['K0','K1','K2','K3'],
    'A':['A0', 'A1', 'A2', 'A3'],
    'B':['B0', 'B1', 'B2', 'B3'],
})

dataset_2 = pd.DataFrame({
    'key':['K0','K1','K2','K3'],
    'A':['A4', 'A5', 'A6', 'A7'],
    'B':['B4','B5','B6','B7'],
})
````

````python
resultado = dataset_1.merge(
    dataset_2,
    how = "inner",
    on = "key"
)
print(resultado)
````

Realiza la misma unión que en el caso anterior, solo que ahora especifica sufijos distintos para columnas con los mismos nombres en ambos datasets.

````python
resultado = dataset_1.merge(
    dataset_2,
    how = "inner",
    on = "key",
    suffixes = ["_df1", "_df2"]
)
print(resultado)
````

### Ejercicio 2

Considera los siguientes dataframes:

```py
dataset_3 = pd.DataFrame(
    {'key1': ['K0', 'K0', 'K1', 'K2'],
    'key2': ['K0', 'K1', 'K0', 'K1'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']}
)

dataset_4 = pd.DataFrame(
    {'key1': ['K0', 'K1', 'K1', 'K2'],
    'key2': ['K0', 'K0', 'K0', 'K0'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']}
)
```

Une los datasets con un *inner join*, usando las llaves key1 y key2 a la vez.

````python
dataset_3 = pd.DataFrame(
    {'key1': ['K0', 'K0', 'K1', 'K2'],
    'key2': ['K0', 'K1', 'K0', 'K1'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']}
)

dataset_4 = pd.DataFrame(
    {'key1': ['K0', 'K1', 'K1', 'K2'],
    'key2': ['K0', 'K0', 'K0', 'K0'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']}
)
````

````python
resultado = dataset_3.merge(
    dataset_4,
    how = "inner",
    on = ["key1", "key2"]
)
print(resultado)
````

Repite la unión anterior pero ahora utiliza un *outer join*.

````python
resultado = dataset_3.merge(
    dataset_4,
    how = "outer",
    on = ["key1", "key2"]
)
print(resultado)
````

Ahora usa un *right join*.

````python
resultado = dataset_3.merge(
    dataset_4,
    how = "right",
    on = ["key1", "key2"]
)
print(resultado)
````

## Aplicación de funciones

### Ejercicio 1

Ejecuta el siguiente codigo para crear un dataset de forma aleatoria.

```py
import pandas as pd
import numpy as np

np.random.seed(42)

# Fechas
fechas = pd.date_range(start="2023-01-01", periods=100, freq="D")

# Datos simulados
ventas = np.random.randint(100, 500, size=100)
gastos_marketing = np.random.randint(50, 200, size=100)
gastos_personal = np.random.randint(80, 300, size=100)

# Crear el DataFrame
df = pd.DataFrame({
    "fecha": fechas,
    "ventas": ventas,
    "gastos_Marketing": gastos_marketing,
    "gastos_Personal": gastos_personal
})

print(df.sample(10)) 
```

````python
import pandas as pd
import numpy as np

np.random.seed(123)

# Fechas
fechas = pd.date_range(start="2023-01-01", periods=100, freq="D")

# Datos simulados
ventas = np.random.randint(100, 700, size=100)
gastos_marketing = np.random.randint(50, 200, size=100)
gastos_personal = np.random.randint(80, 300, size=100)

# Crear el DataFrame
df = pd.DataFrame({
    "fecha": fechas,
    "ventas": ventas,
    "gastos_marketing": gastos_marketing,
    "gastos_personal": gastos_personal
})

print(df.sample(10))
````

Crea una nueva columna **estado** en la que indique si "hay ganancia" (ventas >= gastos_marketing + gastos_personal) o "hay pérdida". Utiliza el método `apply`.

````python
def definir_estado (row):
    if row["ventas"] > row["gastos_marketing"] + row["gastos_personal"]:
        return "hay ganancia"
    else:
        return "hay pérdida"

df["estado"] = df.apply(definir_estado, axis = 1)
print(df.sample(10))
````

### Ejercicio 2

Crea una nueva columna llamada **razon** en la cual se especifique "marketing" si los gastos de este tipo son mayores a los de personal o "personal" en caso contrario, pero solamente cuando el estado sea de pérdida.

````python
def definir_razon (row):
    if row["estado"] == "hay pérdida":
        if row["gastos_marketing"] > row["gastos_personal"]:
            return "marketing"
        else:
            return "personal"
    else:
        return "No aplica"

df["razon"] = df.apply(definir_razon, axis = 1)
print(df.sample(10))
````

### Ejercicio 3

Si las ventas superan el valor de 500 en un día se tiene un buen día. Crea una columna de tipo buleano que refleje este hecho usando el método `apply`.

````python
def definir_dia_bueno (x):
    if x > 500:
        return True
    else:
        return False

df["es_dia_bueno"] = df["ventas"].apply(definir_dia_bueno)
print(df.sample(10))
````

## Visualización con gráficos

### Ejercicio 1

Tomando como referencia el dataset del apartado anterior, crea un gráfico de linea con la evolución de ventas y que tenga varios atributos de forma para hacerlo presentable. Añade además una línea horizonal en el valor de 500 para distinguir los días "buenos".

````python
import matplotlib.pyplot as plt

df.plot(
    y = "ventas",
    x = "fecha",
    figsize = [10,4],
    legend = None,
    title = "Evolución de ventas",
    ylabel = "Ventas",
    xlabel = "",
    marker = "."
)
plt.grid(axis = "y")
plt.axhline(y = 500, color = "red")
plt.show()
````

### Ejercicio 2

Crea un histograma que muestre la distribución de gastos en marketing y los gastos de personal. Recuerda incluir atributos de forma.

````python
df.plot(
    kind = "hist",
    y = ["gastos_marketing", "gastos_personal"],
    bins = 20,
    alpha = 0.5,
    title = "Distribución de gastos",
    ylabel = "Frecuencia observada",
    figsize = [10,4]
)
plt.legend(ncol = 2)
plt.show()
````

### Ejercicio 3

Crea un gráfico de cajón para tener una idea general de la distribución, dispersión y valores atipicos en las ventas, gastos de marketing y gastos de personal. Recuerda incluir atributos de forma.

````python
df.plot(
    kind = "box",
    y = ["ventas", "gastos_marketing", "gastos_personal"],
    figsize = [6,4],
    vert = False,
    title = "Distribución de ventas y gastos",
    color = dict(boxes="lightblue", whiskers="dodgerblue", medians="darkblue", caps="slateblue"),
    patch_artist = True
)
plt.grid(axis = "x")
plt.show()
````
