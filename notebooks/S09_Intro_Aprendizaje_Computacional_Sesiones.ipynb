En el siguiente caso de estudio iniciaremos una primera aproximación al campo del aprendizaje computacional, abordando algunos de los conceptos fundamentales relacionados con la preparación de datos, la construcción de algoritmos y el desarrollo de modelos capaces de extraer conocimiento a partir de información histórica. Esta disciplina constituye una de las áreas más importantes dentro de la ciencia de datos y la inteligencia artificial, ya que permite construir sistemas capaces de identificar patrones, realizar predicciones y apoyar procesos de toma de decisiones de manera automatizada.

Hasta este punto del proceso de aprendizaje, hemos trabajado principalmente en la obtención, limpieza, transformación y análisis exploratorio de datos. Sin embargo, una vez que los datos han sido preparados y comprendidos adecuadamente, para un científico de datos surge la necesidad de utilizar la información disponible para anticipar comportamientos futuros o descubrir estructuras ocultas dentro de los datos. El aprendizaje computacional busca justamente responder a esta interrogante mediante la construcción de modelos matemáticos capaces de aprender a partir de ejemplos y generar resultados útiles frente a nuevas situaciones.

En este sentido, los modelos de aprendizaje computacional pueden perseguir distintos objetivos dependiendo del problema que se desea resolver. Entre los enfoques más comunes se encuentran los siguientes:

* Predicción de variables objetivo a partir de atributos explicativos: el propósito aquí consiste en utilizar información histórica para estimar o predecir valores futuros o desconocidos. Para ello se dispone de un conjunto de observaciones donde se conoce tanto el resultado de interés (variable objetivo) como las variables que potencialmente lo explican (atributos o variables predictoras). El algoritmo analiza estas relaciones y aprende patrones que posteriormente puede utilizar para realizar predicciones sobre nuevos casos. Este tipo de problemas corresponde a los denominados **modelos supervisados**.
* Descubrimiento de patrones ocultos dentro de los datos: en otros escenarios no existe una variable objetivo claramente definida. El interés radica en identificar estructuras, agrupamientos, similitudes o comportamientos que no son evidentes a simple vista. En estos casos, los algoritmos trabajan únicamente con las variables disponibles para encontrar regularidades o segmentos de interés dentro de la información. Este tipo de problemas corresponde a los llamados **modelos no supervisados**.

Como puedes notar, la diferencia fundamental entre ambos enfoques radica en la existencia o no de una variable objetivo conocida durante el proceso de aprendizaje del algoritmo (más comúnmente conocdi como **entrenamiento**). Mientras los modelos supervisados aprenden una relación entre atributos y resultados, los modelos no supervisados buscan comprender la estructura interna de los datos sin disponer de una respuesta previamente definida. En cualquiera de estos escenarios, los datos constituyen el recurso fundamental sobre el cual se construyen los modelos. La calidad de las predicciones o descubrimientos obtenidos depende directamente de la calidad, cantidad y representatividad de la información disponible. Por esta razón, una parte considerable del trabajo en proyectos de aprendizaje computacional se concentra en actividades de preparación de datos, tales como limpieza, transformación de variables, tratamiento de valores faltantes, detección de inconsistencias y selección de atributos relevantes.

Es importante mencionar que los modelos supervisados y no supervisados no constituyen las únicas categorías existentes dentro de la ciencia de datos moderna. Existen otros enfoques cuyo funcionamiento no depende exclusivamente de conjuntos de datos etiquetados, sino también de mecanismos alternativos de aprendizaje y representación del conocimiento. Entre ellos destacan los modelos de aprendizaje por refuerzo, donde un agente aprende mediante interacción con un entorno a través de recompensas y penalizaciones; y los sistemas basados en razonamiento aproximado, que buscan representar y manipular conocimiento bajo condiciones de incertidumbre o información incompleta. Estos enfoques poseen aplicaciones importantes en robótica, optimización, sistemas autónomos e inteligencia artificial avanzada, aunque su estudio excede los objetivos de este caso introductorio.

Volviendo al ámbito de los modelos supervisados con los que iniciaremos nuestra exploración del aprendizaje computacional, es posible distinguir dos grandes categorías dependiendo de la naturaleza de la variable objetivo que se desea predecir:

* Modelos de regresión: cuando la variable objetivo corresponde a un valor numérico, ya sea entero o decimal, el problema se tipifica como un problema de **regresión**. El objetivo consiste en estimar una cantidad o magnitud a partir de un conjunto de atributos explicativos. Algunos ejemplos comunes incluyen pronosticar ventas futuras de una empresa, estimar el precio de una vivienda, predecir la demanda de un producto, calcular el consumo energético de una instalación, o estimar el tiempo necesario para completar una tarea. En todos estos casos, el resultado esperado es un número y el modelo busca aproximarse lo mejor posible a dicho valor.
* Modelos de clasificación: cuando la variable objetivo corresponde a categorías o etiquetas no numéricas, el problema se tipifica como un problema de **clasificación**. En este contexto, el objetivo consiste en asignar cada observación a una clase o categoría específica. Algunos ejemplos frecuentes son determinar si un cliente abandonará o no un servicio, clasificar correos electrónicos como spam o no spam, identificar si una transacción es fraudulenta, diagnosticar la presencia o ausencia de una enfermedad, o clasificar imágenes según el objeto que contienen. En estos casos, el resultado no es un valor numérico entero o decimal, sino una categoría o grupo al que pertenece cada observación.

Durante este caso de estudio nos concentraremos en comprender los fundamentos necesarios para preparar datos y construir nuestro primer modelo supervisado de clasificación, sentando las bases conceptuales que posteriormente permitirán abordar técnicas cada vez más avanzadas. El objetivo principal no será únicamente aprender a ejecutar algoritmos en **Python**, sino comprender la lógica que existe detrás de ellos, la importancia de los datos sobre los cuales se entrenan y los criterios que permiten evaluar si un modelo realmente está generando conocimiento útil para resolver un problema determinado.

## Ingeniería de datos

La ingeniería de datos hace referencia a una serie de actividades que permiten "traducir" los datos a un lenguaje comprensible para los algoritmos por implementar en el modelo de aprendizaje. Para poder cumplir con esto, inicia separando los atributos de las variable objetivo.

````python
X = df_cancer.drop(columns = "tumor_type")
y = df_cancer["tumor_type"]
````

Ahora bien, considera la siguiente idea:

*Puedes imaginar que las computadoras son como niños aprendiendo a sumar. Deben hacer tareas y ejercicios en clase que sean comprensibles pero desafiantes. Luego deben confirmar su aprendizaje en un examen final sin ningún tipo de asistencia de compañeros o maestros. Si superan esta evaluación, se puede asegurar que han aprendido el concepto de suma.*

Por lo expuesto, el primer paso para crear un modelo de aprendizaje computacional consiste en dividir los datos en un subconjunto de **entrenamiento** y otro de **prueba** (o **validación**). Con el primero se busca que el algormitmo a implementar aprenda y se ajuste a las variables existentes. Volviendo al simil del niño aprendiendo a sumar, el conjunto de entrenamiento consiste entonces en las tareas y ejercicios en clase, y el de prueba es el examen final.

Divide entonces el dataset en estas dos partes utilizando la función `train_test_split` de **Scikit-Learn**, tal que el conjunto de prueba represente el 30% de registros originales. ## Creación del modelo base

A continuación vamos a entrenar un algoritmo para que sea capaz de predecir asertivamente si un tumor es maligno o benigno en base a los datos existentes. Para esto, se va a utilizar uno de los algoritmos básicos más populares para resolver problemas de **clasificación**: el *Árbol de decisión*.

Este algoritmo funciona generando nodos y ramificaciones de los datos basadas en reglas de decisión que se derivan de las características observadas en los atributos, y que en su etapa final (hoja del árbol) sean capaces de cumplir su objetivo de predecir la variable objetivo al agrupar en dicha etapa un porcentaje representativo de observaciones con las misma cualidad. En este sentido, cada nodo de decisión interno representa una condición o **parámetro** sobre un atributo (i.e concavity > 0.1), cada rama representa una respuesta posible (True / False), y cada hoja representa un valor de predicción (M o B). 

Una de las principales ventajas de este algoritmo es que es fácil de interpretar y visualizar. También entre sus bondades radica que no exige de muchas condiciones iniciales respecto a los atributos, por lo que simplifica enormemente la ingeniería de datos. Todo esto hace del árbol de decisión uno de los algoritmos más populares y de uso más extendido. 

Sin embargo, y si bien por ahora puede ser dificil de comprender, vale mencionar que una desventaja importante de este algoritmo es que requiere de mucho control en su entrenamiento y optimización, pues de no hacerlo es sensible a generar **sobreajuste**, una condición que perjudicará el rendimiento de nuestro modelo al entrar en producción.

### Modelo base con algoritmo de árbol de decisión

Crea entonces un modelo a partir de este algoritmo usando la función `DecisionTreeClassifier` de **Scikit-Learn**, y establece los argumentos `criterion="gini"` y `max_depth = 5`. No te preocupes, muy pronto vamos a explicar que quieren decir estos términos.

````python
mod_base = DecisionTreeClassifier(
    criterion = "gini",
    max_depth = 5, 
    random_state = 123 # Este argumento asegura la replicabilidad de los resultados del script
)
````

Los argumentos *criterion* y *max_depth* hacen referencia respectivamente al criterio numérico para la creación de ramificaciones y al tamaño máximo del árbol que queremos crear. En términos más técnicos de aprendizaje computacional a estos argumentos lo llamamos **hiperparámetros**, y a diferencia de un **parámetro** normal, estos pueden ser definidos y controlados *a-priori* por el científico de datos. Por consiguiente, son **OPTIMIZABLES**. 

### Entrenamiento del modelo base

Como siguiente paso, has que el modelo base aprenda a pronosticar el tipo de tumor, utilizando el conjunto de entrenamiento (`X_train` e `y_train`) como argumento del método `fit`.

````python
mod_base.fit(X_train, y_train)
````

Como se indicó antes, el algoritmo de árbol de decisión es fácil de visualizar a interpretar. Utiliza la función `plot_tree` para darnos una idea de cómo se ve nuestro modelo base. como argumentos a esta función considera como mínimo lo siguiente:

| Argumento | Valor |
| ---- | ---- |
| decision_tree | La variable que contiene el modelo |
| class_names | El atributo `classes_` del modelo |
| feature_names | El atributo `feature_names_in_` del modelo |

````python
plt.figure(figsize=(18,10))
plot_tree(
    decision_tree = mod_base,
    class_names = mod_base.classes_,
    feature_names = mod_base.feature_names_in_, 
    filled = True, 
    fontsize = 8
)
plt.show()
````

Notemos algunas cosas interesantes de esta visualización:

* El modelo base dice que los atributos concave_points, area, y texture son los más relevantes a la hora de decidir si un tumor es maligno o benigno. Una primera confirmación parcial de nuestra hipótesis al menos en cuanto a las dos primeras variables.
* En cada nodo se evidencia el indice de Gini que hemos establecido como método de creación de ramas, y que mide el grado de igualdad entre las clases B y M de la variable objetivo en cada nodo. Un valor cercano a 0.5 indica que existe una mayor igualdad (hay una misma cantidad de un tipo que de otro), y uno cercano a 0 establece que existe desigualdad (hay más de un tipo que de otro). En consecuencia, es una buena noticia que el indice de Gini en gran parte de las hojas sean en su mayoría valores de aproximadamente 0.  

Confirma la importancia de las variables mencionadas visualizando el porcentaje de relevancia que cada atributo tendrá para el pronóstico del modelo. Para extraer un valor de "importancia" puedes extraer el atributo `feature_importances_` del modelo.

````python
plt.figure(figsize=(6,4))
plt.barh(
    y = mod_base.feature_names_in_,
    width = mod_base.feature_importances_,
    height = 0.8,
    color = "dodgerblue"
)
plt.xlabel("% Importance")
plt.grid(axis = "x", linestyle = ":")
plt.show()
````

### Evaluación del modelo base

¿Qué tan bien predice nuestro modelo base? Para responder a esta pregunta vale enfocarse en dos aspectos importantes que determinan una buena capacidad predictiva:

* El grado de asertividad en el "examen final" debe ser suficientemente alto.
* El grado de asertividad en el "examen final" debe ser similar a aquel de los "ejercicios en clase". 

Veamos el primer aspecto. Construye una **matriz de confusión** que permita contrastar los valores reales de tumor_type con los que predice el algoritmo entrenado. Para esto utiliza el método `predict` con los atributos de prueba (`X_test`) como argumento para generar predicciones con el modelo. 

````python
y_pred = mod_base.predict(X_test)
````

Ahora utiliza la función `crosstab` de **Pandas** ubicando el resultado anterior como columnas y la variable objetivo de prueba (`y_test`) como filas.

````python
mat_conf = pd.crosstab(
    index = y_test,
    columns = y_pred,
    margins = True,
    rownames = ["Real"],
    colnames = ["Predicho"]
)
print(mat_conf)
````

En esta matriz de confusión se evidencia que 99 casos que en efecto eran tumores benignos fueron correctamente pronosticados por el modelo; así también 57 casos de tumores malignos fueron debidamente predichos. Es decir, 156 datos de 171 fueron predichos acertivamente, o lo que lo mismo, el modelo base tiene una **exactitud** del $91.2\%\,(= 156/171)$. 

Concentrémonos también en el tipo de tumor más sensible: el maligno. Fijémonos que 61 casos fueron predichos por el modelo en esta clase, y de ellos 57 efectivamente son tumores malignos. Esto es, el modelo base tiene una **precisión** respecto al tipo de tumor maligno del $93.4\%\,(=57/61)$.

Veamos ahora otra perspectiva pero aún considerando el caso de tumores malignos que nos interesa. En los datos de prueba 68 casos era de esta clase M, y como ya se mencionó 57 fueron bien pronosticados. Entonces, el modelo base tiene una **sensibilidad (recall)** respecto al tipo de tumor maligno del $83.8\%\,(=57/68)$.

Visto esto, muestra un resumen de las métricas de evaluación del modelo utilizando la función `metrics.classification_report` para verificar estos cálculos realizados.

````python
print(metrics.classification_report(y_test, y_pred, digits = 3))
````

Además de los indicadores antes mencionados, en este reporte aparece también la métrica **F1**, que consiste en un promedio armónico de la precisión y la sensibilidad:

$$ F1 = \frac{2 \times precision \times recall}{precision + recall} $$

Ahora bien, notemos que para evaluar un modelo existen diversas métricas a considerar, y el seleccionar una u otra como referencia principal para concluir que un resultado es asertivo dependerá del contexto de cada caso. Dado lo anterior, puede ser conveniente enfocarnos tanto en la **precisión** como en la **sensibilidad** de la clase de tumores malignos, puesto que sería crítico que nuestro modelo no sea capaz de detectar estos tumores cuando de hecho los hay en un paciente (baja sensibilidad), pero igualmente sería contraproducente que el modelo pronostique que un paciente está enfermo, cuando de hecho no lo está (baja precisión). Por tanto, usar el **F1** al ser un promedio de ambas metricas puede resultar útil.

Guarda por tanto esta métrica en una variable utilizando la función `metrics.f1_score` con el argumento `pos_label = M`.

````python
f1_base_test = metrics.f1_score(y_test, y_pred, pos_label = "M")
print(f"F1 para casos malignos en casos de prueba: {f1_base_test:.1%}")
````

Guarda también las otras métricas que hemos visto utilizando las funciones `metrics.accuracy_score`, `metrics.precision_score` y `metrics.recall_score`.

````python
acc_base_test = metrics.accuracy_score(y_test, y_pred)
p_base_test = metrics.precision_score(y_test, y_pred, pos_label = "M")
r_base_test = metrics.recall_score(y_test, y_pred, pos_label = "M")
print(f"Exactitud en casos de prueba: {acc_base_test:.1%}")
print(f"Precision para casos malignos en casos de prueba: {p_base_test:.1%}")
print(f"Sensibilidad para casos malignos en casos de prueba: {r_base_test:.1%}")
````

Cambiemos ahora nuestro enfoque al segundo aspecto mencionado al inicio de esta sección, viendo qué tan similar es el grado de asertividad en el conjunto de prueba con aquel que obtendríamos en el conjunto de entrenamiento. Calcula por tanto la métrica F1 para casos de tumores malignos en el conjunto de entrenamiento y mira la diferencia existente entre el valor obtenido antes. Recuerda que para esto debes hacer una predicción con el método `predict`.

````python
# Predecir con datos de entrenamiento
y_pred = mod_base.predict(X_train)

# Calcular f1 en conjunto de entrenamiento
f1_base_train = metrics.f1_score(y_train, y_pred, pos_label = "M")

# Mirar diferencia entre metrica de prueba y entrenamiento
print(f"F1 para casos malignos en casos de entrenamiento: {f1_base_train:.1%}")
print(f"F1 para casos malignos en casos de prueba: {f1_base_test:.1%}")
print(f"Diferencia observada: {abs(f1_base_train - f1_base_test):.1%}")
````

A esta diferencia se la conoce como **sobreajuste** del modelo y mide que tan bien (o mal) el mismo el capaz de pronosticar cuando desconoce el resultado real. Se podría entender entonces como una medida de la confiabilidad del modelo cuando este ya se encuentre funcionando en un contexto real. Al respecto, es lógico que siempre exista un grado de sobreajuste en los modelos, pues si pensamos en nuestro simil del niño aprendiendo a sumar, diversos factores como el estrés, los nervios o el cansacio, tienden a incidir en que la calificación de su examen final sea levemente menor al de sus tareas y ejercicios en clase.

En nuestro caso de estudio un nivel de sobreajuste de poco más del 10% en valores referenciales de 90 puntos porcentuales no resulta tan preocupante, aunque siempre que puede reducirse sería mejor.

## Optimización del modelo base

En esta etapa nuestro objetivo es maximizar los resultados alcanzados con el modelo base a través de técnicas que modifiquen para mejor el algoritmo utilizado de *Árbol de decisión*. Vamos entonces a conocer e implementar la primera de estas técnicas: el **ajuste de hiperparámetros**. 

Como ya mencionamos, el algoritmo de árbol de decisión tiene como hiperparámetros al tamaño máximo del árbol representado por el argumento `max_depth`, y el criterio de ramificación representado por el argumento `criterion`; y éstos pueden ser controlados por nosotros *a-priori*. 

Define entonces una función `fun_eval_modelo` que permita probar distintos algoritmos en base al método numérico de creación de ramas y al tamaño de árbol, y que devuelva como resultado la métrica F1 para casos malignos en prueba y el sobreajuste.

````python
def fun_eval_modelo (crit, tam):

    # Crear modelo
    modelo = DecisionTreeClassifier(
        criterion = crit, 
        max_depth = tam, 
        random_state = 123
    )

    # Ajustar modelo a datos
    modelo.fit(X_train, y_train)

    # Generar prediccion en entrenamiento
    y_pred = modelo.predict(X_train)

    # Calcular f1 en entrenamiento
    f1_train = metrics.f1_score(y_train, y_pred, pos_label = "M")

    # Generar prediccion en prueba
    y_pred = modelo.predict(X_test)

    # Calcular f1 en prueba
    f1_test = metrics.f1_score(y_test, y_pred, pos_label = "M")

    # Calcular sobreajuste
    sobreajuste = abs(f1_train - f1_test)

    return f1_test, sobreajuste
````

Mediante un bucle `for`, comprueba distintas alternativas de los hiperparámetros a fin de seleccionar la opción que tenga un sobreajuste menor al 10% y a su vez maximice el F1 de casos malignos en prueba. Considera para esto que los criterios de ramificación posibles son: gini, entropy y log_loss.

````python
# Definir tamaños de arbol a probar
tams = range(1, X_train.shape[1] + 1)
criterios = ["gini", "entropy", "log_loss"]

# Definir umbrales iniciales
of_umbral = 0.1
f1_best = 0
tam_best = 0

# Ejecutar optimizacion
for c in criterios:
    for t in tams:

        # Usar funcion de evaluacion
        f1, of = fun_eval_modelo(crit = c, tam = t)

        #Seleccion del mejor modelo
        if of <= of_umbral and f1 >= f1_best:
            of_alcanzado = of
            f1_best = f1
            tam_best = t
            crit_best = c

print(f"Profundidad óptima: {tam_best}")
print(f"Criterio óptimo: {crit_best}")
print(f"F1 alcanzado en prueba: {f1_best:.1%}")
print(f"Sobreajuste alcanzado: {of_alcanzado:.1%}")
````

Con un tamaño de árbol de 2 y un método de creación de ramas "log_loss" hemos mejorado el sobreajuste, además de incrementar la métrica F1. Incluso, puede ser que tengamos mejores resultados en las otras métricas como un efecto positivo adicional.

Para tu conocimiento, el método de Pérdida Logística de Entropía ("log_loss"), al igual que el indice de Gini, mide el grado de similitud en cada nodo, tal que si su valor es cercano a 1 hay mayor igualdad en la cantidad de casos de tipo B y M, y si es cercano a 0 sucede lo contrario. 

Construye y entrena un modelo optimizado con estos hiperparámetros óptimos.

````python
# Crear modelo optimizado
mod_optimo = DecisionTreeClassifier(
    criterion = crit_best,
    max_depth = tam_best, 
    random_state = 123
)

# Entrenar modelo
mod_optimo.fit(X_train, y_train)
````

Genera la matriz de confusión del modelo optimizado con los datos de prueba.

````python
# Predecir con datos de prueba
y_pred = mod_optimo.predict(X_test)

# Generar matriz de confusion
mat_conf = pd.crosstab(
    index = y_test,
    columns = y_pred,
    margins = True,
    rownames = ["Real"],
    colnames = ["Predicho"]
)
print(mat_conf)
````

Muestra un reporte de las metricas alcanzadas con el modelo optimizado en el conjunto de prueba.

````python
print(metrics.classification_report(y_test, y_pred, digits = 3))
````

Compara estos resultados con los obtenidos en el modelo base. Para esto debes calcular primero estas métricas del modelo optimizado.
