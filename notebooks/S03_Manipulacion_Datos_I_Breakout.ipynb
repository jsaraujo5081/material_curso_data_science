## Preparación de los datos

### Tarea 1

Ajusta los nombres de las columnas para que se encuentren en formato *snake_case*.

```python
df_airbnb.columns = [col.lower().replace(" ","_") for col in df_airbnb.columns]
df_airbnb.info()
```

### Tarea 2

Identifica el porcentaje de valores faltantes existentes en cada una de las columnas del dataset, respecto al total de filas.

```python
valores_faltantes = df_airbnb.isnull().sum()
num_filas = df_airbnb.shape[0]
print(100 * valores_faltantes / num_filas)
```

Da tratamiento a valores perdidos conforme los siguientes criterios:

* Para la columna minimum_nights imputa el valor de la mediana.
* Para la columna neighbourhood_group imputa el valor de la moda.
* Elimina las filas con valores perdidos en las columnas price y construction_year.    

```python
#Imputar mediana en minimun_nights
mediana = df_airbnb["minimum_nights"].median()
df_airbnb["minimum_nights"] = df_airbnb["minimum_nights"].fillna(mediana)
```

```python
#Imputar moda en neighbourhood_group
moda = df_airbnb["neighbourhood_group"].mode()[0]
df_airbnb["neighbourhood_group"] = df_airbnb["neighbourhood_group"].fillna(moda)
```

```python
#Eliminar filas con valores perdidos en price y construction_year
df_airbnb = df_airbnb.dropna(subset=["price", "construction_year"])
```

Para el resto de columnas con valores perdidos:

* Si la columna es numérica sustituye los valores faltantes por 0.
* Si la columna es de tipo texto sustituye los valores por la palabra "unknown".

```python
for col in df_airbnb.columns:
    if df_airbnb[col].dtype == "object":
        df_airbnb[col] = df_airbnb[col].fillna("unknown")
    elif df_airbnb[col].dtype in ["int64", "float64"]:
        df_airbnb[col] = df_airbnb[col].fillna(0)
```

Comprueba si aún quedan valores perdidos en el dataset.

```python
valores_faltantes = df_airbnb.isnull().sum()
num_filas = df_airbnb.shape[0]
print(100 * valores_faltantes / num_filas)
```

### Tarea 3

Cambia el tipo de variable de la columna construction_year a entero.

```python
df_airbnb["construction_year"] = df_airbnb["construction_year"].astype(int)
```

## Análisis de datos

### Tarea 1

Muestra cuántas propiedades existen por año de construcción y ordena tu resultado de forma descendente por dicho año.

```python
prop_year = df_airbnb.groupby("construction_year")["id"].count().sort_index(ascending=False)
print(prop_year)
```

Muestra cuántas propiedades existen por grupos de barrios y ordena tu resultado de forma descendente por el conteo.

```python
prop_nb = df_airbnb.groupby("neighbourhood_group")['id'].count().sort_values(ascending=False)
print(prop_nb)
```

### Tarea 2

Muestra cuál es la calificación promedio de clientes por tipo de propiedad. Solamente considera aquellos casos con al menos 5 opiniones.

```python
df_filtered = df_airbnb[df_airbnb["number_of_reviews"] >= 5]
rate_type = df_filtered.groupby("room_type")["review_rate_number"].mean()
print(rate_type)
```

Repite el ejercicio anterior pero ahora presenta los resultados detallando los mismos por tipo de propiedad y grupo de barrio. Excluye también aquellos grupos de barrios que están "mal escritos".

```python
grupos = ["Manhattan", "Brooklyn", "Queens", "Bronx", "Staten Island"]
df_filtered = df_filtered.query("neighbourhood_group in @grupos")
rate_type_group = df_filtered.pivot_table(
    index = "room_type",
    columns = "neighbourhood_group",
    values = "review_rate_number",
    aggfunc = "mean"
)
print(rate_type_group)
```
