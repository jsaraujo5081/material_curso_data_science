# Sprint 8 - Manejo de Bases de Datos con SQL (Sesiones)

El propósito de los siguientes ejercicios es brindarte una introducción en el uso de herramientas para la administración y manipulación de bases de datos, así como en el aprendizaje de la sintaxis fundamental **SQL**, el lenguaje estándar utilizado para interactuar con sistemas de gestión de bases de datos relacionales. A través de estas actividades, se busca desarrollar una comprensión inicial sobre cómo almacenar, organizar, consultar y recuperar información de manera estructurada, integrando también estas capacidades con herramientas de programación y análisis de datos como **Python**.

En el contexto actual, las bases de datos constituyen uno de los componentes más importantes dentro de prácticamente cualquier sistema de información. Empresas, plataformas digitales, aplicaciones móviles, sistemas financieros y proyectos de analítica almacenan grandes volúmenes de datos que requieren mecanismos eficientes para su administración y consulta. Por esta razón, el dominio básico de bases de datos y del lenguaje **SQL** representa una habilidad esencial tanto para desarrolladores de software como para analistas y científicos de datos.

A diferencia de los archivos tabulares tradicionales, las bases de datos permiten gestionar información de manera más robusta, segura y escalable, facilitando el almacenamiento de millones de registros y el acceso eficiente a información relacionada entre múltiples tablas. Comprender cómo funcionan estos sistemas permite desarrollar soluciones capaces de manipular grandes volúmenes de datos de forma organizada y optimizada.

En estos ejercicios trabajarás utilizando **PostgreSQL**, uno de los sistemas de bases de datos relacionales más utilizados a nivel profesional debido a su estabilidad, flexibilidad y potencia para manejar proyectos de distinta complejidad. Asimismo, integrarás este entorno con **VS Code**, permitiéndote administrar bases de datos y ejecutar consultas desde la misma aplicación en la que harás tus análisis y modelos basados en datos.

Inicialmente, los ejercicios abordarán cómo establecer una la conexión entre **PostgreSQL** y **VS Code**, comprendiendo cómo establecer comunicación entre el motor de base de datos y el entorno de trabajo. Esta integración permitirá ejecutar consultas, administrar tablas y trabajar con información almacenada de manera más eficiente y organizada.

Posteriormente, se introducirá el proceso de creación y administración básica de bases de datos, incluyendo conceptos relacionados con tablas, campos de datos, tipos de datos y estructuras relacionales. Aprenderás por tanto cómo diseñar estructuras simples de almacenamiento de información y cómo organizar datos de manera lógica y consistente.

Uno de los componentes centrales de este aprendizaje será la realización de consultas mediante **SQL**. A través de este lenguaje, será posible seleccionar, filtrar, ordenar y transformar información almacenada dentro de las bases de datos que crees. El dominio de consultas constituye una de las competencias más importantes en áreas relacionadas con análisis de datos, inteligencia de negocios y desarrollo de software, ya que permite extraer información útil para análisis, reportes y procesos de toma de decisiones.

Adicionalmente, los ejercicios explorarán la integración entre bases de datos y **Python**, permitiendo ejecutar consultas directamente desde programas desarrollados en **Python**. Esta conexión representa una práctica ampliamente utilizada en proyectos reales de ciencia de datos y analítica, donde frecuentemente se requiere automatizar procesos de extracción, transformación y análisis de información almacenada en bases de datos corporativas.

En conjunto, estas actividades buscan proporcionarte una primera aproximación integral al trabajo con bases de datos relacionales, combinando conocimientos de administración, consultas y programación. De esta manera, podrás comenzar a construir las bases necesarias para enfrentar proyectos más avanzados relacionados con ingeniería de datos, analítica empresarial, automatización y desarrollo de aplicaciones basadas en información estructurada.

### Cálculos 

Has un resumen de la información de visitas web donde sea posible conocer:

* Cuántas visitas se hicieron.
* Cuántas urls se visitaron.
* Cuántos usuarios existieron (un usuario se puede interpretar como una ip única).
* Cuanto tiempo de navegación total se dio en minutos.
* Cuánto tiempo en promedio estuvo un usuario en el dominio (en minutos). 

Utiliza funciones como `count` y `sum`.

```sql
SELECT 
count(id_visitas) AS qvisitas,
count(DISTINCT url) AS qurls,
count(DISTINCT direccion_ip) AS qusuarios,
sum(duracion_seg)/60 AS duracion_total,
(sum(duracion_seg)/60) / count(DISTINCT direccion_ip) AS duracion_prom
FROM fac_visitasweb;
```

Muestra el nombre completo de los empleados que pertenecen al área Comercial o de Finanzas. Asegúrate que el nombre completo tenga el nombre y apellido unido. Emplea para esto la función `concat`.  

```sql
SELECT
concat(nombre,' ',apellido) AS nombre_completo
FROM cat_empleados
WHERE 
departamento IN ('Comercial','Finanzas');
```

Muestra el nombre completo, usuario de email y los años de antiguedad que tiene cada empleado. Para extraer el usuario de email, puedes aplicar la función `split_part`.

```sql
SELECT
concat(nombre,' ',apellido) AS nombre_completo,
split_part(email,'@',1) AS usuario_email,
(CURRENT_DATE - fecha_ingreso) / 365 AS tiempo_en_empresa
FROM cat_empleados;
```

### Agrupaciones

Has un listado de los usuarios web indicando además las visitas que tuvo cada uno y su tiempo total de navegación en minutos. Para agregar la información emplea el comando `GROUP BY`.

```sql
SELECT
direccion_ip,
count(id_visitas) AS qvisitas,
sum(duracion_seg) / 60 AS duracion_total
FROM fac_visitasweb
GROUP BY 
direccion_ip;
```

Modifica el código anterior para que solamente muestre aquellos usuarios que tuvieron más de 30 visitas pero menos de media hora de tiempo de navegacion. Si vas a filtrar información de una consulta que ya ha agregado datos, debes usar el comando `HAVING`.

```sql
SELECT
direccion_ip,
count(id_visitas) AS qvisitas,
sum(duracion_seg) / 60 AS duracion_total
FROM fac_visitasweb
GROUP BY 
direccion_ip
HAVING 
count(id_visitas) > 30 AND sum(duracion_seg) / 60 < 30;
```

### Unión de tablas

Has un listado del top 10 de países que más visitas web han mantenido. Dado que debes extraer información de dos tablas distintas, has una unión entre ellas mediante los comandos `LEFT JOIN` y `ON`.

```sql
SELECT
pais,
count(id_visitas) AS qvisitas
FROM fac_visitasweb AS l
LEFT JOIN cat_ip AS r 
ON l.direccion_ip = r.ip
GROUP BY 
pais
ORDER BY 
qvisitas DESC
LIMIT 10;
```

De los dos países de mayor cantidad de visitas encontrados en el ejercicio anterior, muestra el orden descendente de preferencia las categorías de urls.

```sql
SELECT
pais,
categoria,
count(id_visitas) as qvisitas
FROM fac_visitasweb AS l
INNER JOIN cat_ip AS r1 
ON l.direccion_ip = r1.ip
INNER JOIN cat_url AS r2 
ON l.url = r2.url
GROUP BY 
pais, 
categoria
HAVING 
pais = 'Estados Unidos' OR pais = 'Ecuador'
ORDER BY 
pais,
qvisitas DESC;
```

Muestra una consolidación de toda la información de las tablas **fac_visitasweb**, **cat_ip** y **cat_url**. Dado que la tabla de visitas web y la del catalogo de urls contienen ambas una columna llamada "url", has una subconsulta de esta segunda para cambiar el nombre de este campo a "direccion_url".

```sql
SELECT 
*
FROM fac_visitasweb AS l
LEFT JOIN cat_ip AS r1 
ON l.direccion_ip = r1.ip
LEFT JOIN (
    SELECT 
    url AS direccion_url, 
    categoria, 
    categoria_blog 
    FROM cat_url
) AS r2 
ON l.url = r2.direccion_url;
```

### Creación de vistas

Una vista se puede entender como un esquema tabular de los datos provenientes de tablas en una base de datos, los cuales han sido procesados previamente mediante consultas específicas, y que se guardan como parte de la misma base de datos. Por lo general, estas vistas suelen ser aquellas consultadas por los usuarios de la base de datos como son los analistas o científicos de datos, garantizando así la fidelidad y calidad de las fuentes originales (tablas).

Crea entonces vistas para todas las consultas hechas en la sección anterior con el comando `CREATE VIEW`.

```sql
--Crear vista para consulta de top 10 paises
CREATE VIEW vw_top_paises AS (
    SELECT
    pais,
    count(id_visitas) AS qvisitas
    FROM fac_visitasweb AS l
    LEFT JOIN cat_ip AS r 
    ON l.direccion_ip = r.ip
    GROUP BY 
    pais
    ORDER BY 
    qvisitas DESC
    LIMIT 10
);

--Crear vista para consulta de categorias preferentes
CREATE VIEW vw_categorias_EC_US AS (
    SELECT
    pais,
    categoria,
    count(id_visitas) as qvisitas
    FROM fac_visitasweb AS l
    INNER JOIN cat_ip AS r1 
    ON l.direccion_ip = r1.ip
    INNER JOIN cat_url AS r2 
    ON l.url = r2.url
    GROUP BY 
    pais, 
    categoria
    HAVING 
    pais = 'Estados Unidos' OR pais = 'Ecuador'
    ORDER BY 
    pais,
    qvisitas DESC
);

--Crear vista para consulta consolidada
CREATE VIEW vw_consolidada AS (
    SELECT *
    FROM fac_visitasweb AS l
    LEFT JOIN cat_ip AS r1 
    ON l.direccion_ip = r1.ip
    LEFT JOIN (
        SELECT 
        url AS direccion_url, 
        categoria, 
        categoria_blog 
        FROM cat_url
    ) AS r2 
    ON l.url = r2.direccion_url
);
```

```sql
    salario,
    contrato_permanente,
    (CURRENT_DATE - fecha_ingreso) / 365 AS tiempo_en_empresa
    FROM cat_empleados
);
```

## Consultas de bases de datos con Python

Veamos finalmente cómo podemos consultar a nuestra base de datos desde Python. Abre un nuevo jupyter notebook en VS Code, carga **pandas** y las siguientes librerías que te ayudarán a crear una conexión desde tu notebook con la base de datos en tu servidor de **PostgreSQL**:

* **psycopg2**: Incluye funciones y métodos especializados en la conexión a servidores y bases de datos de **PostgreSQL**
* **sqlalchemy**: Incluye funciones y métodos para la conexión a servidores y bases de datos de diferentes entornos de gestión de bases de datos.

Si no tienes estas librerías aún deberás instalarlas desde tu **terminal** con el comando `pip install`.

````python
import pandas as pd
import matplotlib.pyplot as plt
import psycopg2
import sqlalchemy

import warnings
warnings.filterwarnings(action = "ignore")
````

### Usando psycopg2 para importar información de base de datos

Genera la conexión con la base de datos con la función `connect` (Aquí deberás ingresar los mismos parámetros que cuando te conectaste con VS Code).

````python
conexion = psycopg2.connect(
    host = "localhost",
    port = "5432",
    database = "db_grupoxx",
    user = "postgres",
    password = "123456" #Aqui debes colocar tu propia contraseña
)
````

Escribe una consulta a modo de texto para extraer toda la información de la vista de empleados que creaste y guárdala en la variable `query`.

````python
query = """
SELECT 
* 
FROM vw_empleados;
"""
````

Carga los datos en `df_empleados` desde la base de datos utilizando la consulta y la conexión creada. Puedes emplear la función `read_sql`de **pandas** con el correspondiente argumento `con`.

````python
df_empleados = pd.read_sql(query, con = conexion)
````

Muestra la estructura general de este dataframe para verificar que la información se cargó correctamente.

````python
df_empleados.info()
````

Cierra la conexión creada para garantizar la seguridad de tu base de datos con el método `close`.

````python
conexion.close()
````

### Usando sqlalchemy para importar información de base de datos

Al igual que antes crea una conexion con la base de datos. Para esto, primero define un texto de conexión que incluya los parámetros de conexión de la siguiente forma:

*postgresql://`user`:`password`@`host`:`port`/`database`*

A continuación utiliza la función `create_engine` con este texto como argumento.

````python
# Crear texto de conexion a base de datos
db_config = {
    "host": "localhost", 
    "port": 5432,
    "db": "db_grupoxx",
    "user": "postgres", 
    "pwd": "123456"
} 
texto_con = "postgresql://{}:{}@{}:{}/{}".format(
    db_config["user"], 
    db_config["pwd"], 
    db_config["host"], 
    db_config["port"], 
    db_config["db"]
)

# Generar conexion
conexion = sqlalchemy.create_engine(texto_con)
````

Crea una consulta para extraer la información de la vista consolidada correspondiente a la primera quincena del mes.

````python
query = """
SELECT 
* 
FROM vw_consolidada
WHERE dia <= 15
"""
````

Carga los datos desde la base de datos y guárdalos en `df_consolidada`.

````python
df_consolidada = pd.read_sql(query, con = conexion)
````

Imprime una muestra del dataframe para verificar que la carga fue correcta.

````python
df_consolidada.sample(10)
````

Genera una visualización simple del número de visitas por día.

````python
df_consolidada.groupby("dia").agg(visitas = ("id_visitas","count")).plot()
plt.show()
````

Recuerda cerrar la conexión para terminar el proceso. Utiliza el método `dispose`.

````python
conexion.dispose()
````
