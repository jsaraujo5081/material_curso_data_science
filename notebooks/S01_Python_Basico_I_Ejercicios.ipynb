### Ejercicio 5

Descifra el siguiente mensaje en código usando métodos y funcione de **Python**:

*sgn1rts@3d@0su@l3@n3@0tr3px3@nu@rar3d1sn0c@s3d3up@3t@ay*

````python
mensaje = "sgn1rts@3d@0su@l3@n3@0tr3px3@nu@rar3d1sn0c@s3d3up@3t@ay"

#Reemplazar @ por espacios
mensaje = mensaje.replace("@"," ")

#Reemplazar 3 por e
mensaje = mensaje.replace("3","e")

#Reemplazar 1 por i
mensaje = mensaje.replace("1","i")

#Reemplazar 0 por o
mensaje = mensaje.replace("0","o")

#Invertir el mensaje
mensaje = mensaje[::-1]

print(mensaje)
````

### Ejercicio 1

Crea una lista `years` que contenga los años entre 2014 y 2025, e imprime su longitud.

````python
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]
print(len(years))
````

Crea una lista `dividendos` que contenga 12 números reales cualquiera entre 1 y 3, y que representan los dividendos de la empresa DataCorp en los años generados en la lista anterior. Imprime el tipo de variable de `dividendos`.

````python
dividendos = [1.2, 1.35, 1.5, 1.6, 1.8, 1.95, 2, 2.1, 2.3, 2.5, 1.98, 2.31]
print(type(dividendos))
````

### Ejercicio 2

Calcula la diferencia en los dividendos entre el primer y el último año de la lista, y guarda el resultado en una variable llamada `diferencia`.

````python
diferencia = dividendos[-1] - dividendos[0]
print(diferencia)
````

Crea una nueva lista llamada `dividendos_ultimos` que contenga solo los dividendos de los últimos 5 años.

````python
dividendos_ultimos = dividendos[-5:]
print(dividendos_ultimos)
````

Calcula el promedio simple de los dividendos en los últimos 5 años usando la lista `dividendos_ultimos`. Recuerda que el promedio simple $\mu$ se obtiene de la fórmula:

$$\mu = \frac{\sum x}{n}$$

donde $x$ son los elementos a promediar y $n$ es la cantidad de elementos.

````python
print(sum(dividendos_ultimos) / len(dividendos_ultimos))
````

### Ejercicio 3

Calcula la tasa acumulada de crecimiento de los `dividendos` desde 2014 hasta 2025. Guarda el resultado en la variable `tacc`.

````python
tacc = diferencia / dividendos[0]
print(tacc)
````

Usando el `tacc`, obtén el crecimiento porcentual promedio `p` usando la siguiente fórmula:

$$p = \sqrt[n-1]{1+tacc} - 1$$

````python
p = (1 + tacc)**(1/(len(dividendos) - 1)) - 1
print(p)
````
