## Procesos comerciales

### Tarea 1

Valida si un producto existe en el inventario y si hay suficiente stock para la venta. Para ello, sigue los siguientes pasos:

- Define una función llamada `validar_disponibilidad` que reciba tres parámetros: inventario, product_id y cantidad.
- Verifica si el producto existe en el inventario e informa en el caso de no existir.
- Si existe, verifica si la cantidad solicitada es menor o igual a la disponibilidad.
- Retorna **True** si la validación es exitosa; de lo contrario, muestra un mensaje que informe del caso y retorna **False**.

````python
def validar_disponibilidad (inventario, product_id, cantidad):
    if product_id in inventario:
        if cantidad <= inventario[product_id]["disponibilidad"]:
            return True
        else:
            print(f"Error: No hay suficiente stock de {inventario[product_id]['nombre_producto']}.")
    else:
        print(f"Error: Producto con ID {product_id} no encontrado.")
    return False
````

### Tarea 2

Actualiza la disponibilidad de un producto después de una venta. Para ello, sigue los siguientes pasos:

* Define la función `actualizar_inventario` con tres parámetros: inventario, product_id y cantidad.
* Reduce la disponibilidad del producto en la cantidad vendida.
* Muestra un mensaje con el nuevo stock.

````python
def actualizar_inventario (inventario, product_id, cantidad):
    inventario[product_id]["disponibilidad"] -= cantidad
    print(f"Inventario actualizado: {inventario[product_id]['nombre_producto']} - {inventario[product_id]['disponibilidad']} {inventario[product_id]['unidad_medida']}")
````

### Tarea 3

Calcula el valor de la venta mediante una función llamada `calcular_subtotal`, en la que se devuelva como resultado la multiplicación entre la cantidad comprada y el precio del producto. Considera que deberías acceder a la informacion del precio del producto almacenado en el inventario.

````python
def calcular_subtotal (inventario, product_id, cantidad):
    subtotal = inventario[product_id]["precio"] * cantidad
    return subtotal
````

### Tarea 4

Muestra un resumen de la venta mediante la creación de una función llamada `mostrar_resumen`. Este resumen debería incluir el nombre del producto, la cantidad vendida, el precio unitario y el subtotal vendido, donde todos estos items deberían ser entradas.

````python
def mostrar_resumen(inventario, product_id, cantidad, subtotal):
    print("--- Resumen de la Venta ---")
    print(f"Producto: {inventario[product_id]['nombre_producto']}")
    print(f"Cantidad: {cantidad} {inventario[product_id]['unidad_medida']}")
    print(f"Precio unitario: ${inventario[product_id]['precio']:.2f}")
    print(f"Total: ${subtotal:.2f}")
    print()
````

### Tarea 5

Procesa una lista de compras, validando, calculando y actualizando el inventario. Para ello, sigue los siguientes pasos:

* Define la función `procesa_lista_compras` con dos parámetros: inventario y lista_compras.
* Itera sobre cada producto en la lista de compras.
* Usa las funciones anteriores para validar disponibilidad, calcular subtotales, mostrar resúmenes y actualizar el inventario.
* Suma los subtotales para calcular el total final de la venta e imprímelo.

Para que sirva de guía, a continuación encontrarás unas listas de compra ejemplificativas:

```py
lista_compras_1 = [{"product_id": "101", "cantidad": 5}, {"product_id": "103", "cantidad": 10}]
lista_compras_2 = [{"product_id": "102", "cantidad": 2}, {"product_id": "105", "cantidad": 3}, {"product_id": "110", "cantidad": 1}]
```

````python
def procesar_lista_compras (inventario, lista_compras):
    
    total_venta = 0
    
    for item in lista_compras:
        product_id = item["product_id"]
        cantidad = item["cantidad"]

        if validar_disponibilidad(inventario, product_id, cantidad):
            subtotal = calcular_subtotal(inventario, product_id, cantidad)
            total_venta += subtotal
            mostrar_resumen(inventario, product_id, cantidad, subtotal)
            actualizar_inventario(inventario, product_id, cantidad)

    print(f"Total de la venta: ${total_venta:.2f}")
````
