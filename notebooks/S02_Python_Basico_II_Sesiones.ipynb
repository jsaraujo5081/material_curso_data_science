## Manejo de diccionarios

### Ejercicio 1

## Definición y uso de funciones 

### Ejercicio 1

Escribe una función que tome un nombre cualquiera como entrada e imprima un mensaje de saludo.

### Ejercicio 2

Escribe una función que tome dos edades cualquiera como entrada y devuelva su promedio.

````python
def promedio (edad_1, edad_2):
    prom = (edad_1 + edad_2) / 2
    return prom
````

Utiliza la función con dos edades de clientes de Aleph. Guarda el resultado en una variable llamada `prom_edad` e imprímela.

````python
# Extraer las edades del diccionario
lista_edades = list(cliente_dicc.values()) #La función lista transforma objetos "similares" a listas

#Seleccionar dos elementos cualquiera
edad_ej1 = lista_edades[2]
edad_ej2 = lista_edades[5]

#Llamar a la funcion
prom_edad = promedio(edad_ej1, edad_ej2)
print(prom_edad)
````

### Ejercicio 3

Escribe una función que devuelva True (Verdadero) si un número cualquiera es par, y False en caso contrario.

````python
def es_par (n):
    par = n % 2 == 0
    return par
````

Utiliza la función con dos edades distintas de clientes de forma que un resultado sea Verdadero y otro Falso.

````python
# Extraer las edades del diccionario
lista_edades = list(cliente_dicc.values()) #La función lista transforma objetos "similares" a listas

#Seleccionar dos elementos cualquiera
edad_ej1 = lista_edades[1]
edad_ej2 = lista_edades[3]

#Llamar a la funcion
print(es_par(edad_ej1))
print(es_par(edad_ej2))
````

## Uso de bucles

### Ejercicio 1

Utiliza un bucle `for` para transformar el diccionario de clientes en una lista anidada `cliente_list` donde cada sublista contenga como elementos el nombre y la edad de cada persona. 

````python
#Definir lista vacía
cliente_list = []

for nombre in cliente_dicc.keys():
    caso = [nombre, cliente_dicc[nombre]]
    cliente_list.append(caso)

print(cliente_list)
````

Con la lista obtenida y utilizando un bucle `for`, imprime para cada cliente un mensaje que diga lo siguiente:

*El cliente de nombre (aquí va el nombre del cliente) tiene (aquí va su edad) años.*

````python
for caso in cliente_list:
    print(f"El cliente de nombre {caso[0]} tiene {caso[1]} años.")
````

### Ejercicio 2

Genera un saludo para todos los clientes. (Pista: Utiliza una de las funciones creadas previamente)

````python
#Forma 1 (con la lista)
for caso in cliente_list:
    nombre = caso[0]
    saludar(nombre)
````

````python
#Forma 2 (con el diccionario)
for nombre in cliente_dicc.keys():
    saludar(nombre)
````

### Ejercicio 3

Utiliza un bucle `while` para recorrer la lista de clientes y mostrar la información del primer jubilado (mayor de 65 años) registrado. 

````python
#Definir variable para condición
edad_prueba = 0

#Definir variable para iterar
pos = -1

#Generar bucle
while edad_prueba <= 65:
    pos += 1 #pos = pos + 1
    edad_prueba = cliente_list[pos][1]

print(f"El cliente de nombre {cliente_list[pos][0]} tiene {cliente_list[pos][1]} años.")
````

### Ejercicio 4

Utiliza bucles for y while para crear una lista `clientes_rep` que contenga cada nombre de cliente repetido 3 veces.

````python
#Crear lista vacía para guardar resultados
clientes_rep = []

#Ejecutar bucles
for caso in cliente_list:
    rep = 0
    while rep < 3:
        clientes_rep.append(caso[0])
        rep += 1

print(clientes_rep)
````

## Uso de condicionales

### Ejercicio 1

Recorre el diccionario de clientes y extrae en una lista `clientes_jovenes` todos los nombres de los clientes que tengan menos de 45 años.

````python
#Crear lista vacia para almacenar resultados
clientes_jovenes = []

#Ejecutar bucle para recorrer el diccionario
for nombre in cliente_dicc.keys():
    edad_caso = cliente_dicc[nombre]
    #Guardar nombre si cumple con condición
    if edad_caso < 45:
        clientes_jovenes.append(nombre)

print(clientes_jovenes)
````

### Ejercicio 2

Toma como referencia el código del ejercicio anterior de forma que ahora se guarden las edades de los clientes jóvenes en una lista llamada `edades_jovenes`, y en otra lista `edades_viejos` se guarden las edades de los clientes que no son jóvenes.

````python
#Crear listas vacias para almacenar resultados
edades_jovenes = []
edades_viejos = []

#Ejecutar bucle para recorrer el diccionario
for nombre in cliente_dicc.keys():
    edad_caso = cliente_dicc[nombre]
    #Guardar edades según condición
    if edad_caso < 45:
        edades_jovenes.append(edad_caso)
    else:
        edades_viejos.append(edad_caso)

print(edades_jovenes)
print(edades_viejos)
````

Calcula ahora el promedio de edades de clientes jóvenes y viejos e imprime los resultados con un mensaje adecuado.

````python
prom_jovenes = sum(edades_jovenes) / len(edades_jovenes)
prom_viejos = sum(edades_viejos) / len(edades_viejos)

print(f"El promedio de edad de clientes jóvenes es de {prom_jovenes:.1f} años, mientras que el de clientes viejos es de {prom_viejos:.1f}.")
````

### Ejercicio 3

Aleph busca generar campañas específicas de mercadeo para clientes menores a 30 años, otra para clientes entre 30 y 45 años, y una tercera para clientes cercanos a la jubilación (de igual o más de 45 años). Modifica el código generado en el ejercicio anterior a fin de que el mismo permita contar con listas de nombres de clientes en cada uno de los segmentos relevantes.

````python
#Crear listas vacias para almacenar resultados
nombres_sub30 = []
nombres_sub45 = []
nombres_over45 = []

#Ejecutar bucle para recorrer el diccionario
for nombre in cliente_dicc.keys():
    edad_caso = cliente_dicc[nombre]
    #Guardar edades según condición
    if edad_caso < 30:
        nombres_sub30.append(nombre) 
    elif edad_caso < 45:
        nombres_sub45.append(nombre)
    else:
        nombres_over45.append(nombre)

print("MERCADO 1:")
print(nombres_sub30)
print()
print("MERCADO 2:")
print(nombres_sub45)
print()
print("MERCADO 3:")
print(nombres_over45)
````
