## Creación, acceso y modificación de diccionarios

Los diccionarios son estructuras de datos que permiten almacenar información en pares de *clave* y *valor*, facilitando la organización y recuperación eficiente de datos. A diferencia de las listas, donde los elementos se acceden mediante posiciones numéricas, los diccionarios permiten acceder a la información a través de identificadores específicos (claves), lo que los convierte en herramientas especialmente útiles para representar datos estructurados y relaciones entre elementos. En **Python**, los diccionarios son ampliamente utilizados en tareas de análisis y ciencia de datos, ya que permiten configurar parámetros eficientemente y modelar de información compleja dada su flexibilidad y rapidez de acceso.

## Definición y uso de funciones para automatización de procesos

Las funciones constituyen bloques de código reutilizables diseñados para ejecutar tareas específicas dentro de un programa. Su principal objetivo es modularizar la solución de un problema, permitiendo dividir procesos complejos en partes más pequeñas, organizadas y fáciles de mantener. Mediante el uso de funciones, es posible reducir la repetición de código, mejorar la legibilidad de los programas y facilitar la detección de errores. En **Python**, las funciones pueden recibir parámetros de entrada y retornar resultados, lo que favorece la creación de programas más dinámicos, reutilizables y escalables. Vale mencionar que su uso te permitirá desarrollar procesos de manera más eficiente en distintos ámbitos de la preparación, análisis  y modelamiento de datos.

Escribe una función que tome dos edades cualquiera como entrada y devuelva su promedio aritmético mediante el comando `return`.

```python
def promedio (edad_1, edad_2):
    prom = (edad_1 + edad_2) / 2
    return prom
```

## Creación y uso de bucles o estructuras repetitivas

Los bucles, también conocidos como estructuras repetitivas, permiten ejecutar un conjunto de instrucciones múltiples veces de manera automática. Estas estructuras son fundamentales cuando se requiere procesar colecciones de datos, realizar tareas iterativas o automatizar operaciones repetitivas. El uso adecuado de bucles contribuye significativamente a la eficiencia y simplicidad de un programa, evitando la necesidad de escribir instrucciones redundantes. En **Python**, los bucles más utilizados son los llamados `for` y `while`, cada uno orientado a diferentes tipos de procesos iterativos. Por una parte los bucles `for` facilitan el realizar tareas específicas en cada uno de los elementos de objetos iterables (i.e listas o diccionarios); mientras que los bucles `while` se especializan en repetir procedimientos mientras se cumpla una condición específica.

### Ejercicio 1

Utiliza un bucle `for` para transformar el diccionario de clientes en una lista anidada `cliente_list` donde cada sublista contenga como elementos el nombre y la edad de cada persona. 

```python
# Definir lista vacía
cliente_list = []

# Ejecutar el bucle en cada elemento del diccionario
for nombre in cliente_dicc.keys():
    caso = [nombre, cliente_dicc[nombre]]
    cliente_list.append(caso)

cliente_list
```

Con la lista obtenida y utilizando un bucle `for`, imprime para cada cliente un mensaje que diga lo siguiente:

*El cliente de nombre (aquí va el nombre del cliente) tiene (aquí va su edad) años.*

```python
for cliente in cliente_list:
    print(f"El cliente de nombre {cliente[0]} tiene {cliente[1]} años.")
```

Genera un saludo para todos los clientes utilizando un bucle `for` y una de las funciones creadas previamente.

### Ejercicio 4

Utiliza bucles for y while para crear una lista `clientes_rep` que contenga cada nombre de cliente repetido 3 veces.

```python
# Crear lista vacía para guardar resultados
clientes_rep = []

# Ejecutar bucle for que recorra lista de clientes
for cliente in cliente_list:
    
    # Definir iterable para bucle while
    rep = 1
    
    # Ejecutar bucle while que repita el nombre 3 veces
    while rep <= 3:
        clientes_rep.append(cliente[0])
        rep += 1

clientes_rep
```

## Creación y uso de estructuras condicionales

Las estructuras condicionales permiten que un programa tome decisiones y ejecute diferentes acciones según se cumplan determinadas condiciones lógicas. Este mecanismo constituye la base del control de flujo en programación, ya que posibilita adaptar el comportamiento del programa frente a distintos escenarios o entradas de información. A través de condicionales, un algoritmo puede evaluar expresiones lógicas y seleccionar la ruta de ejecución más adecuada. En **Python**, las instrucciones `if`, `elif` y `else` proporcionan una forma clara y flexible de implementar procesos de decisión dentro de un programa.

### Ejercicio 1

Recorre el diccionario de clientes y extrae en una lista `clientes_jovenes` todos los nombres de los clientes que tengan menos de 45 años.

```python
# Crear lista vacia para almacenar nombres de clientes jovenes
clientes_jovenes = []

# Definir bucle for para recorrer el diccionario
for nombre in cliente_dicc.keys():
    
    edad_cliente = cliente_dicc[nombre]
    
    # Guardar nombre en lista si cumple con condicion mediante un if
    if edad_cliente < 45:
        clientes_jovenes.append(nombre)

print(clientes_jovenes)
```

### Ejercicio 2

Toma como referencia el código del ejercicio anterior de forma que ahora se guarden las edades de los clientes jóvenes en una lista llamada `edades_jovenes`, y en otra lista `edades_viejos` se guarden las edades de los clientes que no son jóvenes.

```python
# Crear listas vacias para almacenar edades de clientes jovenes y "no jovenes"
edades_jovenes = []
edades_viejos = []

# Definir bucle for para recorrer el diccionario
for nombre in cliente_dicc.keys():
    
    edad_cliente = cliente_dicc[nombre]
    
    # Guardar edades en lista si cumplen con condiciones mediante un if y else
    if edad_cliente < 45:
        edades_jovenes.append(edad_cliente)
    else:
        edades_viejos.append(edad_cliente)

print(edades_jovenes)
print(edades_viejos)
```

Calcula ahora el promedio de edades de clientes jóvenes y viejos e imprime los resultados con un mensaje adecuado. Utiliza para esto las funciones `sum` y `len`.

```python
prom_jovenes = sum(edades_jovenes) / len(edades_jovenes)
prom_viejos = sum(edades_viejos) / len(edades_viejos)

print(f"El promedio de edad de clientes jóvenes es de {prom_jovenes:.1f} años, mientras que el de clientes viejos es de {prom_viejos:.1f}.")
```
