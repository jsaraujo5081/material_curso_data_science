# Sprint 17 - Procesamiento de Lenguaje Natural (Sesiones)

Si bien gran parte de las técnicas de análisis y ciencia de datos se desarrollan sobre información estructurada en tablas, donde cada fila representa una observación y cada columna una variable que caracteriza dicha observación, una enorme cantidad de información valiosa se encuentra almacenada en formatos no estructurados. Correos electrónicos, publicaciones en redes sociales, reseñas de clientes, artículos de prensa, documentos legales, mensajes de texto, transcripciones de llamadas y conversaciones en plataformas digitales son solo algunos ejemplos de información que se genera principalmente a través del lenguaje humano. Como consecuencia, las organizaciones suelen disponer de grandes volúmenes de datos que contienen conocimiento potencialmente útil, pero que no pueden ser utilizados directamente por los modelos analíticos tradicionales dada su estructura.

Esta situación representa uno de los mayores desafíos de la analítica moderna. A diferencia de los datos numéricos, el lenguaje humano posee características complejas como contexto, ambigüedad, ironía, sinonimia y múltiples interpretaciones posibles. Las personas somos capaces de comprender estas sutilezas de manera casi natural gracias a nuestra capacidad de razonar, abstraer conceptos y relacionar ideas con experiencias previas. Sin embargo, para una computadora, un texto no es más que una secuencia de caracteres que carece de significado intrínseco. Por esta razón, resulta necesario desarrollar mecanismos que permitan transformar el lenguaje humano en representaciones interpretables por los algoritmos.

El **Procesamiento de Lenaguaje Natural (PLN)** es un campo que surge con el objetivo de abordar este problema, combinando conocimientos de lingüística, estadística, aprendizaje automático e inteligencia artificial para permitir que las computadoras analicen, interpreten y extraigan información útil a partir de textos y otros tipos de comunicación humana. Gracias a los avances en esta área, actualmente es posible desarrollar aplicaciones capaces de clasificar documentos, detectar sentimientos, resumir textos, traducir idiomas, responder preguntas e incluso mantener conversaciones relativamente naturales con las personas.

Desde la perspectiva de la ciencia de datos que estudiamos, una de las principales aplicaciones del **PLN** consiste en convertir información textual en variables numéricas que puedan incorporarse a modelos predictivos. El desafío aquí radica en representar adecuadamente el contenido de un texto preservando la mayor cantidad posible de información útil para el problema que se desea resolver. Para ello, en este caso estudiaremos tres métodos ampliamente utilizados en el **PLN**:

* Codificación por frecuencia: transforma los textos en representaciones numéricas a partir de la cantidad de veces que aparecen determinadas palabras o términos dentro de los documentos.
* Codificación por relevancia implícita: busca identificar qué palabras aportan realmente información diferenciadora dentro de un conjunto de textos, asignando mayor importancia a aquellos términos que ayudan a distinguir unos documentos de otros.
* Codificación por interpretación semántica: utiliza representaciones más avanzadas capaces de capturar relaciones de significado entre palabras y conceptos, permitiendo que los algoritmos trabajen con una noción más cercana al contexto y la semántica del lenguaje humano.

Adicionalmente, aprenderemos diversas técnicas de preparación de textos orientadas a reducir la complejidad de la información antes de su análisis. Estas técnicas permiten limpiar, normalizar y simplificar los textos eliminando elementos poco relevantes, reduciendo ruido y mejorando la calidad de las representaciones que posteriormente serán utilizadas por los modelos predictivos. En la práctica, una adecuada preparación del texto suele tener un impacto tan importante como la elección del algoritmo de aprendizaje computacional.

Finalmente, incorporaremos un nuevo algoritmo dentro de nuestro conjunto de herramientas predictivas: las **Máquinas de Vectores de Soporte (SVM)**. Este algoritmo es uno de los más importantes dentro del aprendizaje supervisado y ha demostrado históricamente un excelente desempeño en problemas relacionados con clasificación de textos. Gracias a su capacidad para encontrar fronteras de separación eficientes entre diferentes categorías, el **SVM** se han convertido en una herramienta ampliamente utilizada en aplicaciones como detección de spam, clasificación temática de documentos y análisis de opiniones.

A lo largo de este caso práctico aprenderemos cómo transformar textos en información utilizable por modelos predictivos, cómo reducir la complejidad inherente al lenguaje humano y cómo aplicar estas representaciones dentro de algoritmos de aprendizaje automático. De esta manera, ampliaremos nuestro alcance más allá de los datos tabulares tradicionales y daremos un primer paso hacia el análisis de una de las fuentes de información más abundantes e importantes del mundo moderno: el lenguaje humano.

## Preparación de datos

Lleva a cabo tu plan de acción para limpiar los textos de los comentarios. Al respecto, puntualicemos en algunas cosas evidentes que se observan en ellos y que deberían ser tratadas:

* Existencia de caracteres en mayúsculas y minúsculas de forma aleatoria.
* Códigos y entidades *html*.
* Símbolos que que no son letras o caracteres, y que por tanto no que forman términos o frases conceptualmente comprensibles en el lenguaje natural (emojis, caracteres especiales, etc.)

Como punto de partida, extrae todos los textos y guárdalos en un nuevo dataset llamado `corpus` para mayor facilidad de procesamiento.

````python
corpus = df_comments[["content"]]
corpus.sample(20, random_state = 123)
````

Crea una nueva columna llamada `content_clean` donde cambies todos los textos de la muestra para que estén en minúscula.

````python
corpus["content_clean"] = corpus["content"].str.lower()
corpus.sample(20, random_state = 123)
````

Sustituye en esta nueva columna los textos tipo código *html* de acuerdo a la siguiente tabla de expresiones regulares:

| Expresión regular | Definición | Caracter por el cual sustituir |
| ---- | ---- | ---- |
| `<\w*\s/>` | Cierre de bloque de código *html* o salto de linea del tipo `<br/>`. | `" "` (Espacio) | 
| `<\w*>` | Inicio de bloque de código *html*. | `" "` (Espacio) |
| `\&#39;` | Representación en código *html* de comilla simple. | `"\'"` (Commilla simple) |
| `\&#*\w*[0-9]*;*` | Otras representaciones de caracteres especiales que no son letras. | `" "` (Espacio) |

Con este propósito, utiliza para cada una la función `re.sub` dentro de métodos `apply`.

````python
# Sustituir cierres de bloque de codigo o saltos de linea
corpus["content_clean"] = corpus["content_clean"].apply(lambda x: re.sub(r"<\w*\s/>"," ",x))

# Sustituir inicios de bloques de codigo
corpus["content_clean"] = corpus["content_clean"].apply(lambda x: re.sub(r"<\w*>"," ",x))

# Sustituir código de comilla simple
corpus["content_clean"] = corpus["content_clean"].apply(lambda x: re.sub(r"\&#39;","\'",x))

# Sustituir códigos de otros caracteres especiales 
corpus["content_clean"] = corpus["content_clean"].apply(lambda x: re.sub(r"\&#*\w*[0-9]*;*"," ",x))
corpus.sample(20, random_state = 123)
````

Sustituye ahora todos los caracteres que no sean de tipo letra, espacios o la comilla simple por un espacio. Esto es, cambia aquellos caracteres que sigan la expresión regular `[^\w\s\']`.

````python
corpus["content_clean"] = corpus["content_clean"].apply(lambda x: re.sub(r"[^\w\s\']"," ",x))
corpus.sample(20, random_state = 123)
````

Dado que hemos quitado diversos caracteres a cambio de espacios, conviene además:

* Eliminar los espacios sobrantes al inicio y al final de texto.
* Eliminar los espacios entre palabras excesivos.

Has estos ajustes en el orden indicado.

````python
# Eliminar espacios sobrantes al inicio y al final del texto
corpus["content_clean"] = corpus["content_clean"].str.strip()

# Eliminar espacios excesivos
corpus["content_clean"] = corpus["content_clean"].apply(lambda x: re.sub(r"\s+"," ",x))
corpus.sample(20, random_state = 123)
````

En estos textos además existen palabras mal escritas, otros códigos extraños que no son *html* y jergas particulares. No nos interesan estos casos por lo que vamos a mantener solamente aquellos términos reconocidos usualmente en el idioma inglés. Para limpiar estos casos, utiliza el diccionario `web2` como argumento de la función `get_english_words_set` de la libreria **english_words**, y únelo con el diccionario `en_core_web_sm` que provee **spacy**. Este último debe instalarse previamente por la terminal con el siguiente comando:

```ps
python -m spacy download en_core_web_sm
```

Asegúrate que este diccionario unificado tenga todas sus palabras en minúsucla e incorpora además los nombres de los artistas que existen en el dataset pues son términos propios de nuestro contexto.

````python
# Extraer y unir diccionarios de terminos en ingles
dicc_gews = get_english_words_set(["web2"], lower = True)
dicc_spacy = spacy.load("en_core_web_sm")
dicc_spacy = set(dicc_spacy.vocab.strings)
dicc_en = dicc_gews.union(dicc_spacy)

# Poner todas las palabras del diccionario en minuscula
dicc_en = {x.lower() for x in dicc_en}

# Añadir nombres de artistas
dicc_en.update({"shakira","eminem","katy","perry","lmfao","psy"})
print("DICCIONARIO INGLES:")
print(list(dicc_en)[:15])
print(f"Cantidad de términos: {len(dicc_en):,.0f}")
````

````python
# Generar funcion para limpiar textos con palabras propias del diccionario 
def fun_palabras_reales (texto):
    palabras = texto.split()
    palabras_reales = []
    for p in palabras:
        try:
            # Incorporar numeros
            palabras_reales.append(str(int(p)))
        except:
            # Incorporar palabras
            if p in dicc_en:
                palabras_reales.append(p)
    return " ".join(palabras_reales)

# Aplicar funcion en textos
corpus["content_clean"] = corpus["content_clean"].apply(fun_palabras_reales)
corpus.sample(20, random_state = 123)
````

Ya tenemos nuestros textos más limpios, no te preocupes si aún encuentras algunas cosas extrañas en ellos, con lo que hicimos es suficiente por ahora. 

Ahora bien, en una nueva columna llamada `content_lemma` estandariza finalmente los textos a través de una **lematización**. Este proceso consiste en eliminar conjugaciones de verbos, y flexiones o declinaciones de sustantivos, a fin de facilitar su transformación posterior. Puedes ayudarte para esto nuevamente con el diccionario `en_core_web_sm` de la librería **spacy**.

````python
# Cargar modelo lematizador
mod_lematizador = spacy.load("en_core_web_sm")

# Lematizar textos
for i in corpus.index:
    texto = mod_lematizador(corpus.loc[i,"content_clean"])
    texto = " ".join([x.lemma_ for x in texto])
    corpus.loc[i,"content_lemma"] = texto.lower()

corpus.sample(20, random_state = 123)
````

### Codificación por frecuencia

Esta codificación parte de una lógica simple (pero no por eso menos conveniente) tal que cada texto se divide en n-gramas o frases de una o más palabras. A partir de allí se hace un simple conteo de veces en que dichos términos se presentan. Tómese por ejemplo la frase 

*"mi nombre es juan cual es tu nombre"*

La codificiación por frecuencia la transformaría en lo siguiente: 

| Término | Frecuencia |
| ---- | ---- |
| "mi" | 1 |
| "nombre" | 2 |
| "es" | 2 |
| "juan" | 1 |
| "cual" | 1 |
| "tu" | 1 |


De esta forma, se da una suerte de ponderación numérica a aquellos términos más comúnmente utilizados en el contexto; y este número puede servir como atributo para ser considerado por un algoritmo en un modelo predictivo.

Vale señalar que resulta conveniente excluir de esta codificación a las llamadas "palabras de para", puesto que las mismas no aportan ninguna información más allá de ser nexos o puentes entre ideas relevantes. Volviendo a nuestro ejemplo, si quitamos los pronombres y preposiciones la transformación quedaría así:

| Término | Frecuencia |
| ---- | ---- |
| "nombre" | 2 |
| "juan" | 1 |
| "cual" | 1 |

Notemos cómo la exclusión de estas palabras da mayor enfoque al contenido relevante del texto. Aplica entonces esta codificación en los textos lematizados. Utiliza la función `CountVectorizer` con su método `fit_transform` e incluye el argumento `stop_words` con las "palabras de para" extraídas.

### Codificación por relevancia implícita

También llamada codificación **TF-IDF**, este procedimiento va un paso más allá respecto al mostrado anteriormente, ya que además de calcular la frecuencia de un término específico (*TF: Text Frecuency*), se incorpora un factor adicional que "castiga" el continuo surgimiento de cada término en el conjunto de textos estudiados (*IDF: Inverse of Document Frecuency*). En este sentido, una palabra que aparece siempre en distintos textos tiene menor relevancia a nivel de cada caso específico, puesto que la información nueva que aporta es menor en el contexto de todo el documento.

La forma de calcular esta relevancia $R_{n,t}$ para un término $t$, que se repite $f_{t,d}$ veces dentro de un texto $d$ que a su vez es parte de un conjunto o corpus $D$, se define a partir de la siguiente fórmula:

$$ R_{t,d} = TF_{t,d} \times \log (IDF_{t,d}) = \frac{f_{t,d}}{\sum_{i\in d} f_{i,d}} \times \log \left( \frac{|D|}{|{d\in D: t\in d}|} \right) $$

Para entender mejor lo que aquí se busca, volvamos a nuestro ejemplo de la frase *"mi nombre es juan cual es tu nombre"*, con su codificación por frecuencia dada por:

| Término | Frecuencia | Frecuencia Relativa (TF) |
| ---- | ---- | ---- |
| "nombre" | 2 | 0.50 |
| "juan" | 1 | 0.25 |
| "cual" | 1 | 0.25 |

Consideremos ahora una respuesta dada por *"hola mi nombre es ana"*, que tendría su propia codificación por frecuencia:

| Término | Frecuencia | Frecuencia Relativa (TF) |
| ---- | ---- | ---- |
| "hola" | 1 | 0.33 |
| "nombre" | 1 | 0.33 |
| "ana" | 1 | 0.33 |

Para nosotros como humanos es claro que ambos textos buscan comunicar los nombres de los emisores; pero además de esto, quisiéramos saber que características adicionales destacan en ambas frases y las distinguen. Calculemos por tanto el componente **IDF** de los términos del cada texto:

| Término | Frecuencia | Frecuencia Relativa (TF) | Inversa de la Frecuencia del Documento (IDF) |
| ---- | ---- | ---- | ---- |
| "nombre" | 2 | 0.50 | 1.00 |
| "juan" | 1 | 0.25 | 2.00 |

### Codificación por interpretación semántica

El desarrollo tecnológico de los últimos años ha permitido contar con diversos modelos capaces de procesar grandes volúmenes de datos a fin de traducir textos con una suerte de comprensión semántica del contenido de información. Estos modelos **LLM (Large Language Models)** han dado cabida al surgimiento de herramientas muy conocidas actualmente como Gemini (Google), ChatGPT/Copilot (OpenAI - Microsoft), Claude (Anthropic), entre muchas otras; las cuales son capaces de interpretar el sentido y contenido de palabras, frases o documentos, prediciendo respuestas coherentes al contexto de las mismas.   

De forma bastante resumida, a continuación puedes visualizar el proceso que utilizan estos modelos para poder realizar sus tareas:

1. Reciben como entrada una consulta a modo de palabra, frase y/o documento.

2. Transforman los términos de esta entrada en elementos conocidos como **tokens**. Para esto, el modelo cuenta con un diccionario propio predefinido que relaciona cada componente de la entrada con un identificador numérico único.

3. Utilizan estos tokens como atributos de un modelo pre-entrenado de **codificación**. Por lo general este algoritmo consiste de una red neuronal compleja llamada **transformador** que es capaz de relacionar semánticamente a cada uno de los tokens recibidos mediante pesos o coeficientes numéricos. A partir de aquí se genera como predicción un vector que contiene **marcadores semánticos** o **incrustaciones** que actúan como una suerte de representación interna del significado percibido por el modelo respecto a esos tokens y al contexto general de la entrada.

4. Utilizan estas incrustaciones como entradas para otra red neuronal de **decodificación**, que traduce estas señales en base a criterios probabilísticos para seleccionar nuevos tokens que mejor complementen la palabra, frase y/o documento recibido inicialmente. 

5. Transforman los tokens definidos en términos "humanos" utilizando nuevamente el diccionario predefinido, y dando así una respuesta a la consulta inicial.

Como puedes evidenciar, un modelo LLM cuenta básicamente con tres estructuras ya establecidas que se usan secuencialmente a lo largo del proceso: diccionario de tokens, algoritmo codificador y algoritmo decodificador. Por lo que si te interesa adentrarte en el funcionamiento en detalle estas estructuras, te recomiendo mirar el siguiente grupo de videos de Youtube: [Deep Learning by 3Blue1Brown](https://www.youtube.com/playlist?list=PLcCe-ymWq77ow42k4-ZrLzlM3F7Ha7smT). 

Volviendo a nuestro proyecto en el que únicamente nos interesa llegar al punto 3 del proceso antes mencionado puesto que estamos **codificando** textos; el modelo que vamos a utilizar es uno desarrollado por Google y que lleva el nombre de **BERT (Bidirectional Encoder Representations from Tranformers)**. El transformador asociado a este modelo fue puesto en marcha en su primera versión el año 2018 y consiste en una red neuronal de 12 capas y más de 100 millones de pesos, capaces de  devolver 768 marcadores semánticos para cualquier entrada que reciban.

## Creación de modelo base

El algoritmo **SVM** es un método de clasificación bastante popular debido principalmente a su eficiencia computacional y a su lógica consistente y robusta. La idea detrás del mismo es que si un conjunto de datos es en principio "separable" en grupos o categorías, entonces es posible construir un hiperplano que los distinga, tal que la distancia de este con la observación más cercana de cada grupo (conocida como margen) se maximice. En las siguiente visualización se aclara de mejor manera este criterio para el caso de dos atributos.

![](https://github.com/jsaraujo5081/material_curso_data_science/blob/main/complementary_files/svm.png?raw=true)

Para efectos de su aplicación práctica, este algoritmo tiene los siguienes hiperparámetros relevantes:

* Gamma ($\gamma$): nivel de sensibilidad del modelo. En otras palabras, que tan granular se desea que el hiperplano sea con los datos existentes. Se suele emplear de base $\gamma = 1/k$, donde $k$ corresponde a la cantidad de atributos existentes.
* C: factor de regularización. Es decir, grado de ajuste deseado en el proceso de maximización del margen al aplicar métodos de gradiente. Se suele emplear de base $C = 1$.

Visto lo anterior, vale ejecutar el modelo con los distintos conjuntos creados y evaluar su rendimiento.

## Optimización de modelo
