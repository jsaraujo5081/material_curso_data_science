## Definición y uso de funciones

### Ejercicio 1

Escribe una función llamada `promedios` que calcule los principales promedios de dos números cualquiera y devuelva los resultados a modo de diccionario. Para esto considerar los siguientes estadísticos:

* Media aritmética ($\bar{x}_a$):

$$ \bar{x}_a = \frac{x_1 + x_2}{2} $$

* Media geométrica ($\bar{x}_g$):

$$ \bar{x}_g = \sqrt{|x_1 \times x_2|} $$

* Media armónica ($\bar{x}_h$):

$$ \bar{x}_h = \frac{2}{\frac{1}{x_1} + \frac{1}{x_2}} = \frac{2\times x_1 \times x_2}{x_1 + x_2} $$

````python
def promedios (x_1, x_2):
    med_a = (x_1 + x_2) / 2
    med_g = abs(x_1 * x_2) ** (0.5)
    med_h = (2 * x_1 * x_2) / (x_1 + x_2)
    dicc = {"media_aritmetica": med_a, "media_geometrica": med_g, "media_armonica": med_h}
    return dicc
````

Comprueba la función creada con dos números cualquiera.

````python
promedios(5.8, 3.7)
````

### Ejercicio 2

Escribe una funcion llamada `varianza` la cual calcule y devuelva este estadístico ($var$) para dos números cualquiera. Para esto considerar la siguiente definición:

$$ var = \frac{(x_1 - \bar{x}_a)^2 + (x_2 - \bar{x}_a)^2}{2} $$

Recuerda que $\bar{x}_a$ es la media aritmética.

````python
def varianza (x_1, x_2): 
    med_a = (x_1 + x_2) / 2
    v = ((x_1 - med_a)**2 + (x_2 - med_a)**2) / 2
    return v
````

Comprueba la función creada con dos de números cualquiera. Guarda tu resultado en una variable llamada `var`.

````python
var = varianza(20.3, 12.1)
print(var)
````

Obtén la raiz cuadrada de `v` y guardale en una variable llamada `desviacion_estandar`.

````python
desviacion_estandar = var**(0.5)
print(desviacion_estandar)
````

## Uso de condicionales

### Ejercicio 1

Escribe una funcion llamada `es_bisiesto` que, dado un año como entrada, devuelva si es bisiesto o no.

Recuerda que para saber si un año es bisiesto, se debe cumplir con las siguientes condiciones:

* El año debe ser divisible por 4.
* Si es divisible por 100, pero NO es divisible para 1000, entonces NO es bisiesto (i.e. 2000 es bisiesto pero 2100 no lo es).

````python
def es_bisiesto (a):
    x = True
    if a % 4 == 0:
        if a % 100 == 0 and a % 1000 != 0:
            x = False
    else:
        x = False
    return x
````

Comprueba si los siguientes años son bisiestos: 96, 100, 1492, 1886, 1990, 2000, 4100.

````python
periodos = [96, 100, 1492, 1886, 1990, 2000, 4100]

for p in periodos:
    print(f"El año {p} es bisiesto: {es_bisiesto(p)}")
````

### Ejercicio 2

Escribe una función `es_primo` que reciba como entrada un número natural cualquiera e imprima un mensaje donde indique si el número es primo o compuesto.

Vale indicar que un número primo es aquel que solamente es divisible por 1 o por sí mismo, por ejemplo el 7, 11, o 13.

````python
def es_primo (n):
    
    primo = True
    
    if n > 2:
        for d in range(2,n):
            if n % d == 0:
                primo = False
                break
    
    if primo:
        print(f"{n} es primo")
    else:
        print(f"{n} es compuesto")
````

Verifica que la función creada funciona para los primeros 50 números enteros. ¿Notas algún patrón?

````python
for n in range(1,51):
    es_primo(n)
````

### Ejercicio 4

La sucesión de Fibonacci es un conjunto de números naturales muy famoso que deberías conocerlo al menos por cultura general. En esta sucesión se cumple que el número en la posición $i$ está dado por $a_i = a_{i-1} + a_{i_2}$, con $a_1 = 0$ y $a_2 = 1$. En este sentido, los primeros elementos de esta sucesión son:

$$ 0, 1, 1, 2, 3, 5, 8, ...$$

Utiliza funciones y condicionales para calcular rápidamente cuál es el 40-número de Fibonacci.


````python
def fibonacci (n):
  a1 = 0
  a2 = 1
  if n == 1:
    resultado = a1
  elif n == 2:
    resultado = a2
  else:
    resultado = fibonacci(n-1) + fibonacci(n-2)
  return resultado
````

````python
print(f"{fibonacci(40):,.0f}")
````
