Sin embargo, detrás de muchas de las técnicas y algoritmos que hemos utilizado existe una base matemática común que resulta fundamental para comprender cómo funcionan realmente los modelos predictivos: el **álgebra lineal**.

El **álgebra lineal** es una rama de las matemáticas dedicada al estudio de vectores, matrices, espacios vectoriales y transformaciones lineales. Aunque estos conceptos pueden parecer abstractos en un primer momento, la realidad es que constituyen el lenguaje matemático sobre el cual se construyen gran parte de las herramientas modernas de análisis y ciencia de datos. De hecho, muchos de los algoritmos utilizados actualmente pueden interpretarse como transformaciones matemáticas aplicadas sobre conjuntos de datos representados en forma matricial.

En este sentido, resulta útil la perspectiva de pensar en un conjunto de datos como una estructura organizada de observaciones y variables. Cada observación puede representarse como un vector de características y el conjunto completo de datos como una matriz. Bajo esta interpretación, cada variable define una dimensión dentro de un espacio vectorial y cada registro corresponde a un punto ubicado dentro de dicho espacio. Esta visión permite comprender de manera más profunda cómo los algoritmos analizan la información y encuentran patrones presentes en los datos.

Muchas de las técnicas que hemos utilizado hasta ahora pueden entenderse precisamente como aplicaciones del álgebra lineal. Por ejemplo, los procesos de escalamiento transforman las coordenadas de los datos dentro de un espacio vectorial; las codificaciones convierten información categórica en representaciones numéricas; y los algoritmos predictivos buscan relaciones matemáticas entre distintas dimensiones de dicho espacio. Aunque en muchos casos estas operaciones se ejecutan mediante librerías especializadas, comprender los principios matemáticos subyacentes permite interpretar mejor los resultados y tomar decisiones más acertadas durante el desarrollo de modelos.

Con el propósito de fortalecer esta comprensión, en este caso abordaremos mediante **álgebra lineal** dos herramientas que ya hemos utilizado previamente: el Análisis de Componentes Principales (PCA) y la Regresión Lineal. Ambos métodos constituyen excelentes ejemplos para comprender cómo las transformaciones matemáticas pueden utilizarse para simplificar datos, reducir dimensionalidad y modelar relaciones entre variables. Más allá de aprender a utilizar estas técnicas, el objetivo será entender qué sucede internamente cuando son aplicadas sobre un conjunto de datos.

Adicionalmente, introduciremos un concepto de gran importancia dentro del modelado predictivo: la **multicolinealidad**. Este fenómeno ocurre cuando dos o más variables explicativas presentan relaciones lineales fuertes entre sí, es decir, cuando contienen información similar o redundante. Aunque a primera vista esto podría parecer inofensivo, la presencia de **multicolinealidad** puede generar dificultades para ciertos algoritmos, especialmente aquellos que buscan estimar relaciones individuales entre atributos y una variable objetivo.

La existencia de atributos altamente correlacionados puede provocar problemas de interpretación, inestabilidad en los parámetros estimados y una menor capacidad de generalización del modelo. En consecuencia, identificar y tratar adecuadamente la **multicolinealidad** constituye una etapa importante dentro del proceso de construcción de modelos predictivos robustos y confiables. De hecho, una de las aplicaciones más conocidas del PCA consiste precisamente en transformar variables correlacionadas en nuevas dimensiones independientes que conservan una parte significativa de la información original.

* Por un lado, dada la existencia de multicolinealidad el modelo será poco generalizable. Un cambio pequeño en uno de los atributos tenderá a incidir en cambios a nivel de todos los demás, alterando el valor predicho de manera poco controlada. En términos matemáticos, nótese que ante **multicolinealidad** de un atributo $x_i$ con el resto, se cumple que:

$$ y = X\cdot w + \epsilon = \sum_{x\in X}(w_x x) + \epsilon $$
$$ \implies \frac{\partial y}{\partial x_i} = w_i + \sum_{x_j \neq x_i}{w_j k_j} $$
$$ \implies \frac{\partial y}{\partial x_i} = w_i + \tau_i \neq w_i $$

donde $\tau_i$ es un ruido en el peso o coeficiente correspondiente que distorciona la predicción deseada.

* Por otro, el modelo no será útil para interpretaciones asociadas al impacto o importancia que tiene un atributo sobre la variable objetivo, lo cual conlleva a que sea menos confiable a la hora de utilizarlo para la toma de decisiones estratégicas.

### Vectores y valores propios

Al igual que un número compuesto puede factorizarse entre sus divisores primos (i.e $24 = 2^3 \times 3$), una matriz puede igualmente descomponerse en factores. La única diferencia radica en que en su caso existe más de una forma de hacerlo. Una de estas maneras es la que utilizaremos a continuación y se denomina **descomposición por vectores propios**. Sea por tanto $A$ una matriz cuadrada de dimensiones $n\times n$, la misma que cumple con

$$ A\cdot v = \lambda\cdot v $$

donde $v$ es un vector de $n$ elementos y $\lambda$ un escalar (número real). Dado que se cumple esta igualdad, se dice que $v$ es un **vector propio** de $A$, y $\lambda$ es su **valor propio** asociado. Por definición existen al menos $n$ vectores propios de $A$, por lo que es factible derivar la descomposición buscada:

$$ A = Q\cdot\Lambda\cdot Q^{-1} $$

donde 
* $Q = [v_1 \quad ... \quad v_n]$ es la matriz compuesta por los $n$ vectores propios de $A$. 
* $\Lambda = diag(\lambda_1 \quad ... \quad \lambda_n)$ es una matriz diagonal con los $n$ valores propios asociados.
* $Q^{-1}$ es la matriz inversa de $Q$ tal que $Q\cdot Q^{-1} = I$ con $I$ siendo la matriz identidad de rango $n$.

Más allá de esta definición, una característica importante de los vectores propios de una matriz es que son **ortogonales**, es decir, son independientes entre sí. Por tanto se cumple que $Q \cdot Q^T = I$ con $Q^T$ siendo la matriz transpuesta de $Q$. De aquí surge una propiedad interesante y es que $Q^T = Q^{-1}$.


Además, y esto quizás te es útil saber en el contexto de nuestro estudio, un término equivalente para hacer referencia a los vectores propios es el de **componentes principales** de la matriz $A$.

### Descomposición de vectores propios para PCA

Ahora bien, ¿cómo se relaciona esto con el **PCA**? Supongamos que mediante este método se desea transformar un conjunto de atributos $X$ que tienen **multicolinealidad** y que están centrados en 0. Para hacerlo, es necesario extraer la matriz de covarianzas $\Sigma_X$ de este conjunto, y aplicando la descomposición de vectores propios antes vista se tiene que

$$ \Sigma_X = Q\cdot \Lambda\cdot Q^{-1} $$

Notemos que dada la independencia entre las columnas de $Q$, toda la información asociada a la **multicolinealidad** existente debería permanecer únicamente en $\Lambda$. Por tanto, si se proyecta $X$ en el espacio de $Q$ (o lo que es lo mismo, si se obtiene el producto punto entre estas dos matrices), se garantiza que el resultado obtenido consista de una matriz con columnas igualmente independientes.

Entonces, la transformación PCA no es más que generar una nueva matriz de $b$ atributos $X_b$ tal que

$$ X_b = X\cdot Q_{(b)} $$

Solamente basta indicar que $b$ hace referencia a la cantidad de componentes principales (o vectores propios) que deseamos considerar.

### Aplicación de PCA paso a paso

Apliquemos todos estos conceptos en nuestros datos para verificar si con esto corregimos la **multicolinealidad** encontrada. 

Primero centra los datos del conjunto $X$ tal que todas las columnas tengan un valor promedio de 0.

````python
X_centrado = X - X.mean()
print(X_centrado.mean().round(2))
````

A continuación genera la matriz de covarianzas de los datos centrados y guárdala en `mat_cov`.

````python
mat_cov = X_centrado.cov()
print(mat_cov.round(2))
````

Extre los valores y vectores propios de esta matriz con la función `np.linalg.eig`, y guarda estos resultados en `l` y `Q` respectivamente. Te sugiero que imprimas estos valores para hacerte una idea de qué son.

````python
l, Q = np.linalg.eig(mat_cov)
print("VECTORES PROPIOS:")
print(Q)
print()
print("VALORES PROPIOS: ")
print(l)
````

Comprueba que $Q$ es ortogonal, o lo que es lo mismo, que los vectores propios son independientes. Utiliza para esto el producto punto de matrices con el operador `@` y el método `transpose`.

````python
print((Q @ Q.transpose()).round(0))
````

Comprueba ahora que la descomposición de vectores propios funciona. Es decir verifica que $ \Sigma_X = Q\cdot\Lambda\cdot Q^{-1} $ utilizando multiplicación matricial (producto punto).

````python
# Calcular factores matriciales necesarios
Q_inv = Q.transpose() # Igualmente se puede utilizar np.linalg.inv(Q)
L = np.diag(l)

# Multiplicar factores matriciales
desc_vp = Q @ L @ Q_inv

# Verificar igualdad
dif = (desc_vp - mat_cov).round(0)
print(dif)
````

Proyecta la matriz de atributos centrada en el espacio de $Q$ (o lo que es lo mismo, multiplica matricialmente ambas matrices).

````python
X_pca = X_centrado @ Q
X_pca.columns = ["pc" + str(x + 1) for x in range(X_centrado.shape[1])]
X_pca.head()
````

Verifica finalmente que estas nuevas columnas ya no tengan **multicolinealidad** mediante el cálculo de su matriz de covarianzas.

````python
print(X_pca.cov().round(2))
````

## Creación de modelo base

Crea y entrena un modelo base con los datos transformados por **PCA** y utilizando un algoritmo de **regresión lineal** con intercepto. Para esto, ten en cuenta que los pesos o coeficientes generados en una regresión se definen como un vector $w$ tal que

$$ w = (X^T \cdot X)^{-1} \cdot X^T \cdot y $$

````python
# Incorporar atributo de valor constante 1 que represente el intercepto
X_pca_cint = X_pca.copy()
X_pca_cint["intercepto"] = 1
X_pca_cint.info()
````

````python
# Calcular pesos según definicion de algebra lineal
w = np.linalg.inv(X_pca_cint.transpose() @ X_pca_cint) @ X_pca_cint.transpose() @ y
w.index = X_pca_cint.columns
print(w)
````

Evalúa el rendimiento de este modelo, tanto de forma visual como calculando la métrica R2 correspondiente. Para esto ten en cuenta que 

$$ y_{pred} = X \cdot w $$

````python
# Generar predicciones
y_pred = X_pca_cint @ w

# Contrastar predicciones con valores reales
plt.figure(figsize = [6,6])
plt.scatter(
    x = y,
    y = y_pred,
    alpha = 0.75,
    color = "dodgerblue"
)
plt.xlabel("Real"),
plt.ylabel("Prediccion")
plt.grid(linestyle = ":")

plt.plot(y, y, color = "red")
plt.show()
````
