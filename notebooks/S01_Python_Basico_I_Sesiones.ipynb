## Manejo de variables

### Ejercicio 1

John tiene 25 años y su hermano 20. Asigna sus edades a las variables `john_age` y `brother_age`, e imprime en pantalla las variables. 

````python
john_age = 25
brother_age = 20

print(john_age)
print(brother_age)
````

Calcula la diferencia de edad en una nueva variable llamada `diff_ages`.

````python
diff_ages = john_age - brother_age
````

Calcula el promedio de edad en una nueva variable llamada `prom_ages`.

````python
prom_ages = (john_age + brother_age) / 2
````

Muestra en pantalla la siguiente frase:

*John tiene `john_age` años y su hermano tiene `brother_age`. La diferencia de edad entre ellos es `diff_ages` y su promedio es `prom_ages`.*

````python
#Forma 1 (Secuenciar elementos)
print("John tiene ", john_age, " años y su hermano tiene ", brother_age, ". La diferencia de edad entre ellos es ", diff_ages, " y su promedio es ", prom_ages,".", sep = "")
````

````python
#Forma 2 (f - strings)
print(f"John tiene {john_age} años y su hermano tiene {brother_age}. La diferencia de edad entre ellos es {diff_ages} y su promedio es {prom_ages}.")
````

````python
#Forma 3 (format strings)
print("John tiene {} años y su hermano tiene {}. La diferencia de edad entre ellos es {} y su promedio es {}.".format(john_age, brother_age, diff_ages, prom_ages))
````

### Ejercicio 2

A John le encantan los secretos y por eso ha escondido sus claves de acceso a sus bitcoins en alguna parte. Para no olvidar dónde las guardó, creó un acertijo que solo los programadores de Python pueden descifrar. Dejó una frase escrita y varias instrucciones.

La frase dice:

*To my lovely wife I leave my key in the safe. Behind the statue of Dio you have a picture of us both.*

Después dejó algunas instrucciones:

-   A la longitud le restas 72, y luego la repites nuevamente pero que 72 se convierta en 69.
-   Lo que está entre 33 y 35 es la segunda palabra.
-   Si miras desde 44 hasta 38 no serás alguien **negativo**.

¿Dónde están escondidas las claves?

````python
phrase = "To my lovely wife I leave my key in the safe. Behind the statue of Dio you have a picture of us both."

t1 = phrase[len(phrase)-72:len(phrase)-69]
t2 = phrase[33:35]
t3 = phrase[-44:-38]

print(t1,t2,t3)
````
